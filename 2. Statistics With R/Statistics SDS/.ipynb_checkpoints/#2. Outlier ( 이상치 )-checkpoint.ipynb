# Outlier 이상치 
 - 이상치 : 주어진 모형에 의해 잘 설명되지 않는 자료점
 - 잔차를 이용하여 구할 수 있다. 

```r
install.packages("mlbench",repos = "http://cran.us.r-project.org")
```

```r
library(mlbench)
```

```r
Orange <- Orange
str(Orange)
```

```r
aLm <- lm (circumference~age + I(age),data=Orange)  # I(age) 가중치. Inverse
summary(aLm)
```

```r
rstudent(aLm)  # 값이 너무 크거나 작은 값이 이상치이다. 
```

```r
options(repr.plot.width = 4, repr.plot.height=4)
plot(Orange$circumference~Orange$age)
```

```r
plot(aLm)
```

```r
install.packages("car",repos = "http://cran.us.r-project.org")
```

```r
library(car)
```

```r
outlierTest(aLm)  # NA : 데이터는 크거나 작은데 Outlier 는 지금 아니라 확인이 필요하다라고 표현. 
```
