## 파이차트 그리기
 - 최근 30일 동안발생한 지진 데이터 

```r
eq<-read.csv("http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_month.csv")
eq$loc = paste(eq$latitude,eq$longitude,sep=":")
G9<-gvisGeoChart(eq,"loc","depth","mag",options=list(displayMode="Markers",colorAxis="{colors:['purple','red','orange','grey']}"
                                                           ,backgroundColor="lightblue"),chartid="EQ")
plot(G9)
```

```r
display_png(file="4.png") 
```

```r
install.packages("ggmap", repos="http://cran.us.r-project.org")
```

```r
library(ggmap)
```

```r
wifi_new <- read.csv("data/wifi_wgs_euc_kr.csv")
summary(wifi_new)
```

```r
str(wifi_new)
```

```r
cent<-c(mean(wifi_new$LON),mean(wifi_new$LAT))
cent
```

```r
bmap<-ggmap(get_googlemap(center=cent,zoom = 11), extent = "device", base_layer = ggplot(
  wifi_new,aes(x=LON, y=LAT, color = INSTL_DIV)))
```

```r
options(repr.plot.width = 6, repr.plot.height=6)
bmap + geom_point()
```

```r
bmap + geom_density2d() + facet_wrap(~INSTL_DIV)
```
