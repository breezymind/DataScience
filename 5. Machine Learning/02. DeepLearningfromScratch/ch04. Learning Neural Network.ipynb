### 4.2.2 교차 엔트로피 오차 
 - $Cross$ $Entropy$ $error$, $CEE$도 자주 이용한다. 
 - $E = - \sum_k{t_klogy_k}$ (자연로그 e가 밑인 $Log$)
 - 정답에 해당하는 인덱스의 원소만 1이고 나머지는 0 
 - 실질적으로 정답일 때의 추정($t_k$가 1일때의 $y_k$)의 자연로그를 계산하는 식 

```python
def cross_entropy_error(y,t):
    delta = 1e-7
    return -np.sum(t * np.log(y + delta))
```

 - np.log를 계산할때 약간의 delta값을 넣어주는데 log(0)은 마이너스 무한대로 가기 때문에 약값의 값을 더해준다. 
 - 정답 2

 - 기계학습 문제는 훈련 데이터를 사용해 학습.
 - 훈련데이터를 통해 Loss Function 값을 구하고, 그 값을 최대한 줄여주는 매개변수를 구한다.

### 4.2.4 (배치용) Cross Entropy Error
 - y 는 신경망의 출력, t는 정답 레이블. 

```python
def cross_entropy_error(y,t):
    if y.dim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
    
    batch_size = y.shape[0]
    return -np.sum(t * np.log(y))/batch_size
```

### 4.2.5 왜 손실함수를 설정하는가? 

 - 목적 : 정확도를 끌어내는 매개변수 값을 찾는 것. 
 - 손실함수가 어떻게 변하나 (매개변수 값의 변화에 따라서)
 - 가중치 매개변수의 손실 함수의 미분이란 '가중치 매개변수의 값을 아주 조금 변화시켰을 때, 손실함 수가 어떻게 변하나'라는 의미 
 - 만약 미분 값이 음수면, 그 가중치 매개변수를 양의 변수로 변화시켜 손실 함수의 값을 줄일 수 있다. 반대로 양수이면 그 가중치 매개변수의를 음의 방향으로 변화시켜 손실 함수의 값을 줄일 수 있다.

 위 그림과 같이 수치 미분에는 오차가 포함 된다. 
- 오차를 줄이기 위해서 (x+h)와 (x-h)일 때의 함수 $f$의 차분을 계산하는 방법이 쓰인다
- 이 차분은 x를 중심으로 그 전후의 차분을 계산한다는 의미에서 중심 차분 혹은 중앙 차분이라고 한다. ((x+h)와 x의 차분은 전방차분이라고 한다.)

```python
def numerical_diff(f,x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)
```

#### 예 
 - $y=0.01x^2+0.1x$

```python
def function_1(x):
    return 0.01 * x**2 + 0.1 * x
```
