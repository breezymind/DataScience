```python
import pandas as pd
import datetime as dt

df = pd.read_csv('Parking_Tags_Data_2012.csv')
df
```

```python
df.dtypes
```

```python
pd.isnull(df['date_of_infraction']).value_counts()
```

```python
date_format = '%Y%m%d'

def create_weekday(x):
    d = dt.datetime.strptime(str(x['date_of_infraction']), date_format)
    return d.strftime('%A')

df['weekday'] = df.apply(create_weekday, axis=1)
weekday_counts = df['weekday'].value_counts()
weekday_counts
```

```python
weekdays = pd.DataFrame({ 'label': weekday_counts.keys(), 'count': weekday_counts.values })
weekdays
```

```python
weekdays = weekdays.reindex_axis(sorted(weekdays.columns, reverse=True), axis=1)
weekdays
```

```python
weekdays = weekdays.reindex_axis([4,3,1,0,2,5,6], axis=0)
weekdays
```

```python
weekdays.to_csv('weekdays.csv', index=False)
```
