# 탐색적 인자분석 ( Exploratory Factor Analysis )

## 인자분석(요인분석)이란?
 - 심리학, 행동과학 등의 여러 영역에서 주된 관심의 개념을 직접 측정 불가하기 때문에 간접적으로 측정 (랜덤 오차 포함)
  - 예: 지능- 시험점수
  - 예: 사회계급 – 직업, 교육배경, 집의 소유 등
 - 잠재변수 (latent variable) - 직접 측정이 불가하지만 측정할 수 있는 변수
 - 인자분석 – 측정변수와 잠재변수 사이의 관계를 밝히는 것
  - 탐색적 인자분석(Exploratory Factor Analysis) <br> 어떤 측정변수가 어떤 인자에 관련된다는 특정 한 가정 없이 측정변수와 인자 사이의 관계 조사
  - 확인적 인자분석(Confirmatory Factor Analysis) <br> 사전에 가정된 특정한 인자모형에 대해 측정 변수 사이의 공분산 또는 상관관계를 적절하게 적합하는지 검정

## 주성분분석 vs 인자분석  
 - 주성분 분석
  - 변수 20개의 변수를 5개의 축약된 변수로 바꾸는 방법
  - 차원 축소가 주된 의미
 - 인자분석(요인분석)
  - 잠재된 변수를 찾아낸다는 것에 의미가 크다 ( 측정할 수 없는 )

## 인자분석의원리: one-factor model
 - 학생들의 심리테스트
  - 5개의 문항
  - paragraph comprehension (PARA)
  - sentence completion (SENT)
  - word meaning (WORD)
  - addition (ADD)
  - counting dots (DOTS)
 - 5개의 문항에 내재된 하나의 공통 인자(common factor): $f$
 - 각 문항의 특정인자(specific factor): $u_i$ - 랜덤변동
 - 인자분석에서 추정하는 것은 분산 (공통인자의 분산 + 특정인자의 분산)
 - 특정인자의 분산은 순전히 공통인자를 완벽히 측정하지 못하는 데서 오는 오차라고 가정

<img src="photo/1.PNG">

 - 서로간의 상관관계를 보는 것
 - 즉, 서로에게 영향을 주는 요인이 있으니깐 Cor값이 0이 아닌 값이 나온다. 

$$\begin{bmatrix} 
  X_1 = l_1f + u_1 \\
  X_2 = l_2f + u_2 \\
  X_3 = l_3f + u_3 \\
  X_4 = l_4f + u_4 \\
  X_5 = l_5f + u_5
 \end{bmatrix}$$
 <br>
 $$ X : Each \ Measured \ Data $$
 $$ f : Common \ Factors $$
 $$ Specific \ Factor : Individual \ Factors $$

 - 가정
  - $U_i$ 들간의 상호 독립성
  - $U_i$와 $f$ 간의 독립
 - 각 변수 $X_i$의 분산의 요소 : 공통인자$(f)$과 특정인자$(U_i)$ 
  - 둘 다 측정 불가능 
 - $l_i:$인자 적재값 (Factor Loading)
  - 각 변수가 공통인자를 얼마나 반영하는지에 대한 계수
 - $X_i$와 $f$가 표준화 된 경우 <br> $var(X_i) = var(l_if+u_i) = l_i^2 + var(u_i) = 1 $ (표준화 된 값)
  - $l_i = X_i$와 $f$의 상관계수 
  - $l_i^2 : X_i$의 분산 중 공통인자에 의해 발생하는 분산의 비율 $=>$공통성$(Common)$
  - $var(u_i):$특정인자에 의한 분산$(:=\psi_{ii}^2) \ \ u$ 별 $factor$의 분산 
  - $Communality$가 크면(1에 가까우면) 변수 $X_i$가 공통인자를 잘 측정 <br>
    $Communality \ of \ X_i = 1 - \psi_{ii}^2$ <br>
    $Communality \uparrow = \ \ \psi_{ii}^2 \downarrow$
   

## 인자분석의 원리 : Two factor model
 - 학생들의 성적이 두 개의 잠재력의 표현: 언어능력$(f_1)$ , 수리능력$(f_2)$
 - $f_i$들이 서로 독립 가정 
 $$Communality\ of\ X_i = l_{11}^2+l_{12}^2 \gg Var(X_1) = l_{11}^2 + l_{12}^2 + \psi_{ii}^2 $$
 - 높은 언어능력과 낮은 수리능력을 가진 학생
  - $X_2$ $(sentence completion)$의 결과는 언어능력에 의존
  - $l_{21} \thickapprox 1, l_{22}\thickapprox 0$
  
 <img src="photo/2.PNG">

$$ x = Lf+u \gg Cov(x) = LL' + \psi$$
<br>
<center>$ f = \begin{Bmatrix}  f_1 \\ \vdots \\ f_n \end{Bmatrix} \  u = \begin{Bmatrix}  u_1 \\ \vdots \\ u_n \end{Bmatrix} x = \begin{Bmatrix}  x_1 \\ \vdots \\ x_n \end{Bmatrix} \  L = \begin{Bmatrix}  l_{11},l_{12} \\ \vdots \\ l_{n1},l_{n2} \end{Bmatrix} $  </center>

    

## 인자분석 모형
 - $x = Lf+u$
 - $q$개의 측정변수, $k$개의 인자
 - $L = \begin{Bmatrix}  l_{11} \ \cdots \  l_{1k} \\ \vdots \ \ddots \ \vdots \\ l_{q1} \ \cdots \ l_{qk} \end{Bmatrix} , \ f=(f_1,\cdots,f_k)^T,\ u=(u_1,\cdots,u_q)^T $
 - $f$와 $\epsilon$는 서로 독립 
 - $E(\ f) = 0, Cov(\ f) = I $
  - $I = \begin{bmatrix} 1 \ 0 \ 0 \\ 0 \ 1 \ 0 \\ 0 \ 0 \ 1 \end{bmatrix} $ &nbsp; 분산이 1이고 두 $f$가 관계가 없다.
 - $E(u) =0, Cov(u) = \psi, (\psi$는 대각 행렬$)$
  - $\psi = \begin{bmatrix} \psi_2^2 \ 0 \ 0 \\ 0 \ \psi_2^2 \ 0 \\ 0 \ 0 \ \psi_n^2 \end{bmatrix} $

```r
library(psych)
```

```r
applicants <- read.table("Applicant.TXT", header=T)
applicants <- applicants[,-1]
```

```r
summary(applicants)
```

```r
describe(applicants)
```

```r
pairs.panels(applicants)
```

```r
app_s <- scale(applicants)
```

 - 보기가 힘들다 factor1 에 대해 상관관게가 높은 것 부터 보는 것이 좋다.

```r
fa1 <- factanal(app_s,4) # 4개의 factor로 나누어서 인자분석을 한다. 
fa1
```

```r
summary(fa1)
```

 - 인자적재값 행렬$(factor \ loading \ matrix):l_{ij}$
 - 공통성$(communality)=l_{i1}^2+\cdots+l_{ik}^2$
 - $X6.LC$의 공통성 = $0.84^2+0.11^2+0.29^2=0.802$
  - 4개의 공통인자에 의해 $80.2%$ 설명, $19.8%$ 는 특수 인자. 
 - $X5.SC=0.92^2+0.14^2=0.866$
  - $X5\psi_5^2= 1-0.866 = 0.134$
 - $Var(X_1)=1, \ l_1f+Var(u)=1$
 
 - $factor$값에 $-$가 있다면 상관관계가 음의 관계가 있다면 그것에 영향을 미치지 않을까 싶다. 
 - $factor$가 0에 가깝다면 제외시켜준다 ( 빈칸 )
 - $loading$값들이 $l$이 된다. $l_{11}, l_{12},l_{13}, l_{14}$ 

```r
print(fa1,digits=2,sort=T)
```

- factor1에 대해서는 큰 값을 가지고 나머지 factor에 대해서는 작은 값들로 sorting 되어있다. 
- factor1을 보면 총 7개의 변수에 크게 영향을 미쳤네? 
- factor2를 보면 총 3개 정도의 변수에 크게 영향을 미쳤네? 
- 무엇이 영향을 미쳤을까? => 애매함과 주관성이 들어가게 된다. 

                       Factor1 Factor2 Factor3 Factor4
        SS loadings       5.57    2.47    2.10    1.01
        Proportion Var    0.37    0.16    0.14    0.07
        Cumulative Var    0.37    0.54    0.68    0.74

- 인자1: LC(명석), SM(마케팅), DR(추진력), AM(야망), GC(개념파악능력), PO(잠재력)
 - 마케팅능력=(LC+SM+DR+AM+GC+PO)/6
- 4개의 공통인자에의해 원변수변동량의 74%가 설명된다.
- $SS\ loadings$의 $factor1$
 - 얼마나 설명해주는가? 
- $Proportion Var$ 
 - 비율. 
- 전체 분산 합이 1이 안되는 이유는 총 $factor$의 갯수를 최대치로 올리면 된다. 15개
        Test of the hypothesis that 4 factors are sufficient.
        The chi square statistic is 84 on 51 degrees of freedom.
        The p-value is 0.00247 
- 4개의 인자가충분한 가에 대한 $X^2$검정
 - 원 변수가 다변량정규분포를따른다는 가정
 - 최대우도 방법에서 가능 
 - $p-value<0.05 =>$ 충분하지 않다.
- 여기서 귀무가 무엇이냐? $H0 : 4$개의 $factor$가 충분하다. 
 - 즉, 기각. 충분하지 않다. 5개로 늘려봐도 된다.

 - $uniquenesses$ 개별 요인들의 분산 
 - $X14.KJ$의 경우 거의 0에 가깝다. 즉, 개별 요인들의 영향이 적다.
 - $X2.APP$의 경우 가장 개별 변수 영향도가 높다.

### Stock_price.csv는 뉴욕증권시장에 상장된 5개의 주식의 주별 주가수익률 자료이다. 5개 회사의 주가수익률 관계를 통해 주가에 공통적으로 영향을 미치는 잠재적 요소를 찾아내려 한다.
  1.	적절한 공통인자의 수를 구하시오. $2$개
  2.	다양한 Rotation을 적용한 것과 하지 않은 것의 인자적재값을 비교하고 적절하다고 판단되는 결과를 고르시오.
  3.	위에서 결정한 방법을 사용해 각 인자의 이름을 정하시오. 
 - $factor1$ 은행 (금융권) $factor2$ 기름회사(정유)
  4.	각 회사 주가수익률의 공통성(communality)를 구하고 공통인자에 의해 가장 잘 설명되는 회사와 가장 잘 설명되지 않는 회사를 찾으시오.
 - 공통인자에 대해 잘 설명해주는 회사는 RoyalDuchShell 개별인자가 거의 없다. , 개별 인자가 가장 큰 영향을 미치는 회사는 WellsFargo
  5.	각 인자가 설명하는 총 분산의 비율을 구하시오.
 - $factor1$ : 0.34 $factor2$ : 0.30 => 65%
  6.	위에서 구한 인자를 사용해 해당 인자에 대한 인덱스를 생성하고 인덱스 간의 관계를 산점도를 통해 살피시오.
