#### 1. 대략적인 그래프 수치를 확인하여 수치상으로 해결한다. 

```r
options(repr.plot.width=4,repr.plot.height=4)
plot(Crime$robbery,Crime$murder) # robbery 500건 이상 또는 murder 30건 이상의 이상치 발견 데이터 확인 
```

```r
plot(Crime$forcible_rape,Crime$robbery) # robbery 600 건이상과 rocible_rape 80건 이상 의 이상치가 보인다.
```

```r
Crime[Crime$robbery > 600,]    # District of Columbia
Crime[Crime$forcible_rape>80,]  # Alaska
```

#### 2. Outlier Library 활용. 

```r
install.packages("outliers", repos = "http://cran.us.r-project.org")
```

```r
library(outliers)
```

```r
tmpCrime <- Crime
rownames(tmpCrime) <- Crime[,1]
```

```r
tmpCrime <- tmpCrime[,-c(1,9)]
```

```r
head(tmpCrime)
```

```r
out <- outlier(tmpCrime)
out
```

```r
windows()
plot(Crime$robbery,Crime$forcible_rape,xlim=c(0,800))
identify(Crime$robbery,Crime$forcible_rape,labels=Crime$state)
```

```r
windows()
plot(Crime$robbery, Crime$population)
identify(Crime$robbery,Crime$population,labels=Crime$state)
```

```r
options(repr.plot.width=8,repr.plot.height=6)
out=match(c("United States","Alaska","District of Columbia"),Crime$state)
clr=rep(1,dim(Crime)[1])
clr[out]=2
pairs(Crime[,-1],col=clr,pch=clr)
```

```r
Crime2=Crime[-out,]
pairs.panels(Crime2[,-1])
```

```r
symbols(Crime2$murder,Crime2$robbery,circles=Crime2$population)
```
