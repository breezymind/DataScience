# 공분산

```r
x <- 1:100
x1 <- 101:200
df <- data.frame(x=x,x1=x1)
sd(x) # Σ (X - X bar)² / n - 1
cov(df$x,df$x1) # Σ (X - X bar)(X1-X1 bar) / n - 1
```

# 다변량 정규분포 

```r
install.packages("rgl", repos = "http://cran.us.r-project.org")
```

```r
library(mvtnorm)
library(rgl)
```

```r
x2=x1=seq(-4,4,0.1)
x1=matrix(x1,length(x1),length(x1))
x2=matrix(x2,length(x2),length(x2),byrow=T)
```

### 데이터가 cor가 없는 경우 

```r
# independent normal
f=dmvnorm(cbind(as.vector(x1),as.vector(x2)))
f=matrix(f,dim(x1)[1],dim(x1)[1])
```

```r
persp3d(x1,x2,f,col="lightblue",xlab="x1",ylab="x2")
```

```r
library("IRdisplay")
display_png(file="1.png", width = 600)
```

### cor가 가질 경우 
- 찌그러진 종 모양으로 나오지만 각각의 변수 방향으로 봤을때는 정규분포 모양을 가진다. 

```r
# correlated normal
f=dmvnorm(cbind(as.vector(x1),as.vector(x2)),c(0,0),matrix(c(1,0.5,0.5,1),2,2))
f=matrix(f,dim(x1)[1],dim(x1)[1])
```

```r
persp3d(x1,x2,f,col="lightblue",xlab="x1",ylab="x2")
```

```r
library("IRdisplay")
display_png(file="2.png", width = 600)
```
