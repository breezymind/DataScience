## 평균에 대한 추록
 - 추정 : 표본을 통해 모집단 특성을 추측
 - 가설검정 : 모집단 실제 값이 얼마나 되는가 하는 주장과 관련해서 표본이 가지고 있는 정보를 이용해서 가설이 올바른지 판정

## 모집단 평균의 구간 추정
 - 신뢰구간
 - 이 부분의 예제로 mu=가설에서 설정한 평균값   이 필요로 한다. 
 - p-value 값이 0.05보다 크면 귀무가설이 참이다 => 기각하지 않는다. 
 - 실제적으로 우리가 알고 싶은 것은 대립가설 즉, alternative에 있는 조건이 우리가 알고 싶어하는 조건이다. 

```r
library(reshape)
tips <- tips
t99 <- t.test(tips$total_bill, conf.level = 0.99)  # 신뢰 수준 99%
t99
```

## 설명 ↑
 - 신뢰 구간 :  18.30631 ~ 21.26557
 - 신뢰수준 : 99%
 - mean = 19.78594
 - p.value = 3.59580346668899e-96
 - 현재 mu 로 평균값을 지정하지 않았기 때문에 비교 자체가 이상하게 되어 있어서 귀무가설이 기각이 나왔다. 

```r
t90 <- t.test(tips$total_bill, conf.level = 0.90)  # 신뢰 수준 90% 신뢰 수준을 정할 수 있다.
t90
```

## 평균에 대한 추록
 - One-sample T-Test ( 단일 집단 모평균에 대한 검정 ) 

## 예제. 
     예) 2006년조사에의하면한국인의1인1일평균알코올섭취량은
     8.1g이다. 2008년대통령선거로 알코올섭취량이달라졌는지조사하기
     위해10명을무작위로 뽑아서조사한결과다음과같은데이터를얻었다.


```r
# 알콜 섭취량 
OneSample <- c(15.5,11.21,12.67,8.87,12.15,9.88,2.06,14.5,0,4.97)
OneSample
```

     H0 : 2008년의 평균 알콜 섭취량 8.1g
     H1 : 2008년의 평균 알콜 섭취랑 <> 8.1g
     조건 : 정규분포 또는 편중된 그래프 일 경우 표본수가 30~50개 이상. 

## 정규 분포를 따르는지 확인 
 - shapiro.test 

```r
shapiro.test(OneSample) # 정규분포를 따르는지 확인.
```

- W = 0.92341, p-value = 0.3863
- p-value가 0.05보다 크다. 기각X => 정규분포를 따른다. 
- p-value < 0.05 => 관측수가 적어 다른 TEST 고려.

```r
options(repr.plot.width=4, repr.plot.height=3)
qqnorm(OneSample)
qqline(OneSample)
```

### H0 : 𝜇 = 8.1 vs H1 : 𝜇 != 8.1

```r
t.test(OneSample,mu=8.1)
# p.value > 0.05 
# 귀무 가설 참 => 기각 X 평균이 8.1이다.
```

### H0 : 𝜇 = 8.1 vs H1 : 𝜇 > 8.1

```r
t.test(OneSample,mu=8.1,alternative = "greater")
```

## 설명 ↑ 
 - 8.1 이상일 것이라고 판단하여 테스트 해 본 결과 p.value = 0.265 > 0.05 귀무가설이 참 
 - mean 이 9.181 일지라도 평균 8.1 이라는 통계적 범위 내에서 존재한다고 판단. 

### H0 : 𝜇 = 8.1 vs H1 : 𝜇 < 8.1

```r
t.test(OneSample,mu=8.1,alternative = "less")
```

```r
display_png(file="kmu/chapter2/4.PNG", width = 600)  
display_png(file="kmu/chapter2/5.PNG", width = 600)  
```

 - H0 : 암이 걸리지 않았다. H1: 암이 걸렸다.
 - 제 1종 오류, 제 2종 오류 
 - 제 1종 : 참인 가설을 기각. => 암이 걸리지 않았는데 걸렸다고 판정.
 - 제 2종 : 거짓인 가설을 기각 안함. => 암이 걸렸는데 안걸렸다고 판정. 
 - 문제에 따라서 a를 조정할 필요가 있다. 
 - 유의수준(𝛼) 조정

```r
t.test(OneSample,mu=8.1,conf.level = 0.99)
```

## 예제
     15세 이상 관람가 영화의 평균 관객수를 95% 신뢰구간을 통해 추정
     15세 이상 관람가 영화의 평균을 1,500,000 보다 크다고 주장 이 주장이 사실인지 가설 검정을 판단. 
     1. 귀무가설, 대립 가설설정
     2. 가정을 만족하는지 체크
     3. 검정통계량과 p-value 구하기
     4. a = 0.05에서 통계적 결론은 무엇인가.

```r
DF <- read.csv("SLData/movie_MBA2.csv",stringsAsFactors = F)
DF$rating[DF$rating == "12세이상관람가"] <- "12"
DF$rating[DF$rating == "15세이상관람가"] <- "15"
DF$rating[DF$rating == "청소년관람불가"] <- "under19"
DF$rating[DF$rating == "전체관람가"] <- "all"
DF15 <- DF[DF$rating=="15",]
```

```r
options(repr.plot.width=4, repr.plot.height=3)
par(las=1,mfcol=c(1,1))
hist(DF$total_seen,20)
```

```r
str(DF)
```

```r
DFRe <- t.test(DF15$total_seen,conf.level = 0.95,mu=1500000,alternative = "greater")
DFRe
```

     1. 귀무가설 : 영화 평균 1,500,000 이하이다. 
     1. 대립가설 : 영화 평균 1,500,000 보다 크다. 

```r
DFRe$p.value # => 0.02 즉, a보다 작다. H0를 기각 가능하다. 귀무가설이 거짓이다. 1,500,000이다. 
```

```r
DFRe2 <- t.test(DF15$total_seen,conf.level = 0.95,mu=1500000)
DFRe2$conf.int # 150만을 포함하고 있지 않다. 기각한다. 
```

## 평균에 대한 추론 / 두 모집단 평균에 대한 검정. 독립표본 T-검정.

```r
DT <- read.csv("SLData/dental.csv",stringsAsFactors = F)
str(DT)
head(DT)
```

```r
par(mfcol=c(1,2))
options(repr.plot.width=6, repr.plot.height=4)
boxplot(resp~treatment,data=DT)
boxplot(log(resp)~treatment,data=DT)
# 갭차이가 적게 나는 log() 하는 것이 비교 하기 좋다. 
```

     Test가 control군의 분산보다 크다.
     둘다 편중
     Log변환 후 분산 차도 좁혀지고 편향도도 작아졌다.

## 양의 왜도, 음의 왜도가 생길경우는 log 또는 sqrt (루트) 적용. 
 - One sample : T = x bar - 𝜇 / (=sd(x bar)/sqrt(n))
 - Two sample : T = x1 bar - x2 bar / sqrt(Var(x1 bar - x2 bar))
  - sqrt(Var(x1 bar - x2 bar)) => sd(x1)^2 / n1 + sd(x2)^2 / n2
 - 등분산 검정 

```r
display_png(file="kmu/chapter2/10.PNG", width = 600)  
display_png(file="kmu/chapter2/11.PNG", width = 600)  
display_png(file="kmu/chapter2/12.PNG", width = 600)  
```

```r
var.test(log(resp)~treatment, data=DT)  # alternative : true ratio of variances is not equal to 1
# 두개의 분산이 같다. 
```

```r
t.test(log(resp)~treatment,var.equal=T,data=DT)
```

## 설명 ↑ 
 - 결론 p.value=0.0357 < 0.05  : 귀무가설 기각 =>  5%하에서 차이가 있다. 평균이 차이가 있다. 
 - Alternative 는 two difference in means is not equal to 0 이다. 즉, 대립가설이 참이다.

```r
display_png(file="kmu/chapter2/13.PNG", width = 600)  
```

```r
var.test(resp~treatment, data=DT) # 기각
```

    p-value = 0.03483 < 0.5 
    confidence interval: 0.008840233 0.815484912  
    ratio of variances 0.08490628  신뢰구간에 존재 하지 않는다. 

```r
t.test(resp~treatment,data=DT) # 기각
```

## 예제

    # 실전
     3. 12 / 15 세 관객수 비교
     3.	한 영화사에서는 15세 이상 관람가 영화의 평균 관객수가 12세이상 관람가 영화의 평균관객수보다 많다고 주장한다. 
       본 주장이 사실인지 가설검정을 통해 판단하시오 
       i.	귀무가설, 대립가설 설정
       ii.	가정을 만족하는지 체크  
       iii.	검정통계량과 p-value 구하기
       iv.	α=0.05에서 통계적 결론은 무엇인가? 

```r
MV <- read.csv("SLData/movie_MBA2.csv",stringsAsFactors = F)
MV$rating[MV$rating == "12세이상관람가"] <- "12"
MV$rating[MV$rating == "15세이상관람가"] <- "15"
MV$rating[MV$rating == "청소년관람불가"] <- "under19"
MV$rating[MV$rating == "전체관람가"] <- "all"

MV1215 <- MV[MV$rating=="15"|MV$rating=="12",]
```

```r
unique(MV1215$rating)
```

```r
MV1215$rating <- factor(MV1215$rating,labels=c("12","15")) #rating에 원래 다른 levels가 존재한걸 없애준다.
```

```r
options(repr.plot.width=6, repr.plot.height=4)
par(mfrow=c(1,2))
boxplot(total_seen~rating,MV1215)
boxplot(log(total_seen)~rating,MV1215)
```

```r
var.test(total_seen~rating,MV1215)        # p-value = 0.0263  > 귀무 참 
var.test(log(total_seen)~rating,MV1215)  #  p-value = 0.784
```

```r
t.test(total_seen~rating,MV1215,alternative="less")  #12세 - 15세가 0보다 작아야되므로 less를 적용.
#실질적으로는 15세가 크게 나왔지만 통계적으로 유희수준을 넘을 정도로 크게 나오지 않음.
```

```r
t.test(log(total_seen)~rating,MV1215,var.equal = T,alternative="less")
#log는 그냥 해본것. 
```
