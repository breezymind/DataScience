 - 더욱 빠른 탐색적 자료분석을 위한 R 기본 그래프 
 - plot, barplot, hist, boxplot

- plot(독립변수, 종속변수)
- plot(종속변수 ~ 독립변수)

```r
DF <- read.csv("data/example_studentlist.csv")
DF$sex <- as.character(DF$sex)
DF$sex[DF$sex == "남자"] <- "M"
DF$sex[DF$sex == "여자"] <- "W"
DF$sex <- as.factor(DF$sex)
str(DF)
```

```r
plot(DF$age)
```

```r
plot(DF$height,DF$weight) 
```

```r
plot(DF$weight ~ DF$height)
```

```r
plot(DF$height,DF$sex) # 독립변수가 수치형이라 sex (명목형) 변수도 1과 2로 대체하여 나타냄.
```

```r
plot(DF$sex,DF$height) # 독립변수가 Sex(명목형) 이므로 해당 그래프를 boxplot으로변형하여 height를 그려줌
```

### plot(값,값,값) 형태는 안되지만 plot(data.frame) 은 가능하다.

```r
plot(DF[,c(3,7,8)])  
```

 - 각 항목별 상관관계를 볼때 사용되므로 유의해 두어야 한다. 

```r
options(repr.plot.width=10, repr.plot.height=10)
plot(DF)
```

## Levels 별로 그래프 그리기 
 - ex) 몸무게와 키의 산포도 그래프를 보면서 점의 종류로 남자와 여자의 별도로 표시를 원한다면 

```r
options(repr.plot.width=7, repr.plot.height=5)
plot(DF$weight ~ DF$height,pch=as.integer(DF$sex))
legend("topleft",c("M","W"), pch=DF$sex)
```

### coplot() 조건화 그래프 
 - 어떤 명목형 변수에 대해 알아보기 위한 그래프 

```r
options(repr.plot.width=8, repr.plot.height=4)
coplot(DF$weight ~ DF$height|DF$sex)
```

### 그래프 주석
 - 그래프의 정보를 달기위한 명령어
 - grid()
 - abline()
 - title(main : 그래프의 이름  xlab = "독립변수"  ylab = "종속변수")

```r
options(repr.plot.width=5, repr.plot.height=4)
plot(DF$weight~DF$height, ann=F)  # ann=F를 통해 독립, 종속 변수 (라벨) 삭제 
title(main="Relation of height and weight",xlab="height",ylab="weight")  
grid()  # 격자 추가 
heightMean<- mean(DF$height)
abline(v=heightMean, col = "red")  # v = 세로
weightMean <- mean(DF$weight)
abline(h=weightMean, col = "blue")  # h = 가로
```

```r
FreqBlood <- table(DF$bloodtype)
barplot(FreqBlood)
title(main="BloodType Count",xlab="BloodType",ylab="Freq")
```

- tapply(종속변수, 그룹별, 함수)

```r
HeightOfBlood <- tapply(DF$height, DF$bloodtype, mean)
print(HeightOfBlood)
barplot(HeightOfBlood, ylim=c(0,200))  # ylim = y축 길이를 0~200
```

```r
plot(HeightOfBlood) 
```

```r
boxplot(HeightOfBlood)
```

- Levels별로 Boxplot 그리기
 - boxplot(종속변수 ~ 독립변수)

```r
boxplot(DF$height ~ DF$bloodtype)
```

### Histogram
 - hist(데이터, breaks="막대갯수")  # prob=T => 상대도수밀도
 - 한 화면에 여러개의 그래프 그리기 par(mfrow=c(row,colum))

```r
options(repr.plot.width=10, repr.plot.height=5)
par(mfrow=c(1,3)) 
hist(DF$height) 
hist(DF$height,breaks = 10) # 계급갯수를 늘린다. breaks 개 만큼.
hist(DF$height,breaks = 10, prob=T) # 상대도수밀도
lines(density(DF$height))  # hist에 line추가 
```

```r
par(mfrow=c(2,3))    # 2 x 3 개 그래프 그리기. 
plot(DF$weight,DF$height)
plot(DF$sex,DF$height)
barplot(table(DF$bloodtype))
boxplot(DF$height)
boxplot(DF$height~DF$bloodtype)
hist(DF$height,breaks=10)
```

```r
options(repr.plot.width=4, repr.plot.height=4)
par(mfrow=c(1,1))
plot(DF$weight ~ DF$height + DF$age + DF$grade + DF$absence + DF$sex) # 넘어가면서 그래프 보기 
```

### Line 겹쳐 비교하기

```r
TS1 <- c(round(runif(30)*100)) # 랜덤 값 
TS2 <- c(round(runif(30)*100)) # 랜덤 값 

TS1 <- sort(TS1, decreasing = F)
TS2 <- sort(TS2, decreasing = F)

plot(TS1, type = "l")
lines(TS2, lty="dashed",  col="red")
```
