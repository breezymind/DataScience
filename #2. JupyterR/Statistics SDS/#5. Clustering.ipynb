# Clustering ( 군집합 )
 - 패턴공간에 주어진 유한개의 패턴들이 서로 가깝게 모여서 무리를 이루고있는 패턴 집합을 클러스터(군집), <br> 무리짓는 과정을 클러스터링이라 한다.
 - 즉, 다량의 데이터를 유사성이 많은 것끼리 모음으로써 동질성 ↑, 서로 다른 그룹간에 이질성 ↑ 
 - 고차원 데이터를 시각화하거나 압축하는 기법 중 하나 
 - 데이터 분석과 패턴인식을 통해 내재된 클러스터 구조를 추출하는 것.

# Classification vs Clustering 
 - Classification : 목표변수가 존재
 - Clustering : 목표변수가 존재 X 

# Clustering 특징
 - 장점 
  - 대용량 데이터에 대한 탐색적 분석.
  - 분석방법의 적용상의 용이성으로 사전지식을 요구하지 않음.
  - 일반적으로 실행속도가 빠름.
  - 언제나 Cluster가 나눠짐.
 - 단점
  - 언제나 Cluster가 구분되지만 목적에 안맞을 수 있음.
  - 목적에 맞는 변수를 제공해야 해당 목적에 맞게 Cluster이 됨.
  - 결과 해석이 어려움.

```r
data(iris)
niris <- iris
```

```r
niris$Species <- NULL
```

```r
kc <- kmeans(niris,3)  # 3개 종류로 나눈다. 실제로 3가지임을 알기에 선정한 것. 실제로 할 때는 최적의 k값을 선정하는 작업이 필요. 
kc
```

## 비교

```r
table(iris$Species,kc$cluster)
```

```r
options(repr.plot.width=4, repr.plot.height=4)
```

```r
plot(niris[c("Sepal.Length","Sepal.Width")],col=kc$cluster)
points(kc$centers[,c("Sepal.Length","Sepal.Width")], col=1:3, pch=8, cex=2)
```

```r
mydata <- na.omit(iris[,-5]) # na 제외
```

```r
mydata <- scale(mydata)
```

```r
wss <- 0
for(i in 1:15) wss[i] <- sum(kmeans(mydata,centers=i)$withinss)
```

```r
plot(1:15, wss, type="b", xlab="Number of Custers", ylab="Within group sum of squares")
```

# Partitioning  Clustering
 - PAM(Partitioning Around Medoids) : 좀 더 견고한 k-means
 - Missing Data 허용
 - profiling시 실제 관측치로 표현
 - 대용량 데이터 처리 시 성능 저하

```r
library(cluster)
iris.km <- kmeans(iris[,-5],3)
table(iris.km$cluster, iris$Species)
```

```r
iris.pam <- pam(iris[,-5],3)
table(iris.pam$clustering, iris$Species)
```

```r
summary(iris.pam)
```
