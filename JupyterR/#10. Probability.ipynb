# 확률 
 - 실험으로 나올 수 있는 결과를 모두 알고 있지만 어떤 결과가 나올지는 모르는 실험을 '확률실험' 이라고 한다. 
 - 확률을 그래프로 그리기 : 확률분포 
  - 확률밀도함수 : 수학 점수처럼 연속형 변수가 확률 변수 일 때
  - 확률질량함수 : 나이와 같이 이산형 변수가 확률 변수 일 때

## 확률 밀도 함수와 확률 질량 함수의 차이 
 - 가장 큰 차이점 : 연속성 
  - 확률 질량 함수는 표기된 X 값 외에는 사용할 수 없다. 
  - P(0.5) 는 구할 수 있지만 P(0.5423) 는 구할 수 없다. 
  - 하지만 확률 밀도 함수는 X 축의 모든 실수 값을 구할 수 있다. 
 - 확률질량함수 : 독립성이 있다.
 - 확률밀도함수 : 연속성이 있다. 

## 확률의 합
 - 확률 질량 함수의 X축에 따른 Y 값을 모두 더하면 100이다. 
 - 확률 밀도 함수 0~6 까지의 모든 확률 값을 더하면 100%이다. 하지만 P(0) + P(1) + P(2) .... + P(6) = 1 이 될까? 아니다. 
  - 0~6까지의 모든 실수값에 해당 되는 확률 값을 더해야 된다. 즉, <b><font size=4>적분</font></b>을 해야한다.

## 정규분포 그래프
 - 평균중심으로 종모양인 그래프를 정규분포 그래프라고 한다. 

## 중심극한 정리 
 - 모집단은 정규분포를 따를 수도 따르지 않을 수도 있다. 하지마 모집단에서 수차례 값을 뽑게 된다면 ( 표본 ) 평균에 가까운 점수를 받을 가능 성이 높다. 
 - 모집단이 정규분포를 따르든 아니든 상관없이 그곳의 점수를 다시 그래프로 그리면  평균값이 많기 떄문에 그래는 반드시 정규분포 그래프가 된다는 것이 <font size =4 color =red > 중심극한 정리 </font> 이다.

### 확률 계산 하기 
#### 정규분포 확률
 - 평균 30 표준편차 7 인 정규분포에서 확률변수가 15의 확률 
 - 평균 550이고 표준편차가 80인 정규분포에서 확률변수가 450의 확률 

```r
dnorm(15, mean=30, sd=7)
dnorm(450, mean=550, sd=80)
```

#### 이항분포 확률 구하기 
 - 동전을 1000번 던져 490번 앞면이 나올 확률 
 - 흡연율이 25%로 알려진 1020명이 있는 학교에서 50명을 뽑았을때 흡연자일 확률?

```r
dbinom(490,size=1000,prob=0.5) 
dbinom(50,size=1020,prob=0.25)
```

### 구간확률계산하기
 - pnorm()
 - pbinom() 
 - 0부터 원하는 값까지의 면적을 구해 확률을 알려준다. 
 - 구간을 원하면 차이를 구하면 된다. 

#### 확률밀도함수 구간 확률 구하기 
 - mean 240 sd 80   에서  0~235까지의 확률 
 - mean 350 sd 75   에서 200~ 370까지의 확률 

```r
pnorm(235, mean=240, sd=80)
pnorm(370, mean=350, sd=74) - pnorm(200, mean=350, sd=75)
diff(pnorm(c(200,370),mean=350,sd=75))
```

#### 확률질량함수 구간 확률 구하기 
 - 2000번 동전을 던져서 앞면이 0~900번까지 나올 확률
 - 2000명있는 기업에서 기혼률은 78% 입니다. 0~1555명을 무작위로 뽑았을때 모두 기혼자일 확률은?
 - 위 상황에서 300~1500명까지 무작위로 뽑았을 때 모두 기혼자일 확률은?

```r
pbinom(900,size=2000,prob=0.5)
pbinom(1555,size=2000,prob=0.78)
diff(pbinom(c(300,1500),size=2000,prob = 0.78))
```

## 가설 검정

    토익학원 평균 623점 표준편차 90으로 알려져 있고, 새로운 학습법을 통해 평균이 700점이 넘는다고 주장한다. 
    H0 : μ = 623   vs   H1: μ>=700

```r
CI <- qnorm(c(0.05,0.95),mean=623,sd=90)
CI # 신뢰 구간
```

```r
Sample <- read.csv("data/example_test_statistic_ex1_sample.csv")
Sample <- Sample[-1]
head(Sample,3)
```

```r
mean(Sample$point)
sd(Sample$point)
```

```r
shapiro.test(Sample$point)
qqnorm(Sample$point)
qqline(Sample$point)
```

```r
t.test(Sample$point,mu=700) # 평균이 700점이라는 것은 거짓. 기각이다.
```

```r
t.test(Sample$point,mu=700,alternative = "greater") # 평균이 700점이라는 것은 거짓. 기각이다.
#p-value = 1 즉, 귀무가 무조건 맞다. 대립은 거짓 . 700이상이란말은 거짓말.
```
