# Overfitting, Generalization
 - 보유한 전체 데이터 ( 100% )
 - Training ( 70% ) - 참고서 문제집
 - Test ( 30% ) - 모의고사 => 학습이 완료 된 모델이 얼마나 뛰어난 것인지 확인하기 위한. 

## Underfitting, Overfitting 
 - Under => 제대로 학습이 안된 상황.
 - Over => 너무 주어진 데이터만 잘 따라가는 상황. 새로운 점이 나오면 오차가 나올 확률이 높다.
 - 학습 오차가 너무 크거나(학습X) 너무 적어도(Overfitting) 안 좋다.
