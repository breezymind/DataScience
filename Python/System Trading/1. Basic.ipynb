# System trading with Algorithm using Python 

```python
import pandas as pd
import pandas.io.data as web
import datetime
```

```python
def download_stock_data(file_name, company_code, year1,month1,date1, year2,month2, date2):
    start = datetime.datetime(year1,month1,date1)
    end = datetime.datetime(year2,month2,date2)
    df = web.DataReader("%s.KS" % (company_code),"yahoo",start,end)
    df.to_pickle(file_name)
    return df
```

```python
download_stock_data('samsung.data','005930',2016,1,1,2016,8,30)
```

```python
def load_stock_data(file_name):
    df = pd.read_pickle(file_name)
    return df
```

```python
df = load_stock_data('samsung.data')
print df.describe()
```

# 시계열 데이터 
 - 일정 시간 간격으로 측정된, 순서를 갖는 데이터를 일컫는다. 
 - 즉, 시간의 흐름에 따라 값이 변하는 데이터들이 시계열 데이터 => 주가, 온도, 상품판매량, 환율 데이터 등이 있다. 

## 시계열 데이터 분석 
 - 주가 데이터는 시계열 데이터 분석이 매우 중요한 부분을 차지하고 있다.
 - 대표적인 목적
  - 시계열 데이터 패턴에 영향을 미치는 요소(Feature) 를 찾는다.
  - 과거의 데이터가 어떻게 미래의 데이터에 영향을 미치는지 분석
  - 미래 데이터를 예측.

## 주요 시계열 데이터 특성
 - **Trend** :  증가나 감소 또는 반복 등의 일정한 패턴이나 경향을 가지고 있는가?
 - **Seasonality** : 일,월,년 계절 등 일정 시간에 따라 지속해서 반복되는 패턴이 있는가?
 - **Outliers** : 이상치를 관측 할 수 있는가?
 - **Long-run Cycle** : 계절성과는 별도로 오랜 기간 반복되는 패턴이 있는가?
 - **Constant Variance** : 측정값이 일정한 수준 이내로 변동되는가 아니면 변동이 무작위로 발생하는가?
 - **Abrupt Change** : 급격한 변동을 보이는 데이터가 있는가?

## 랜덤과정 
 - 확률변수가 시간의 흐름에 따라 변화한 값 
 - 특정한 값으로 귀결되거나 일정한 패턴을 보이지 않기 때문에 랜덤과정을 시간의 흐름에 따른 확률분포라 생각 할 수 있다. 
 - 임의과정이라고도하며 일정한 규칙 없이 움직이는 **주가와 환율** 같은 것이 대표적인 예. 
 
## 정상 시계열 데이터 
 - 정상성 : 평균과 분산 같은 통계적 특성이 시간에 대해 일정한 성질을 의미 이러한 랜덤 과정을 정상과정이라고 한다. 
 - 주로 비정상성과 정상성을 비교 한다. 
 - 정상성 
  - 약한 정상성(Weak-sense Stationarity) : 평균함수와 공분산이 시간에 대해 변하지 않는 성질
  - 강한 정상성(Strong-sense Stationarity) : 위와 반대

# 공분산
 - Cov(x,y) > 0 : 양의 상관 
 - Cov(x,y) < 0 : 음의 상관 
 - Cov(x,y) = 0 : 관계가 없음. 

```python
download_stock_data('hanmi.data','128940',2016,1,1,2016,8,30)
```

## 자기 공분산
 - 약한 정상성 과정 변수의 공분산 함수
 - C(k) 는 Lag K에 대한 자기공분산함수를 지칭 
 $$ C(k)  = \dfrac{1}{n} \sum_{}^{n-k} (x_{t} - \bar x)(x_{t+1}-\bar x) $$
 - Lag K 는 Xt와 Xt+k 사이의 상관관계를 의미 
 $$ Cov(x,y)  = \dfrac{1}{n-1} \sum_{i=1}^{n} (x_{i} - \bar x)(y_{i}-\bar y) $$
 - 공분산은 동일 시간에 2개 변수의 상관관계를 분석하는 것이라면 자기공분산은 서로 다른 2개의 시간 사이에 대한 변수 값의 공분산을 계산 

## 자기상관 
 - 시계열 변수의 시간에 따른 자기 상관관계를 나타내는 것으로 계열상관 또는 교차자기상관이라고 한다. 
$$ P(k) = \dfrac{C(k)}{C(0)} $$
