```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.hani.co.kr/arti/list1.html"
source_code = requests.get(url)
source_code.encoding = 'utf-8'
plain_text = source_code.text
bsObj = BeautifulSoup(plain_text,"html.parser",from_encoding="utf-8")
```

```python
for link in bsObj.find_all("h4"):
    for link2 in link.find_all("a", {"href":re.compile("\/[0-9]*\.html")}):
        link3 = link2.get("href")
        if (link3.startswith("http://www")):
             url = link3
        else:
             url = "http://www.hani.co.kr" + link3        
        atcl = requests.get(url)
        atcl.encoding = 'utf-8'
        atcl_text = atcl.text
        atclBS = BeautifulSoup(atcl_text, "html.parser",from_encoding="utf-8")
        for hit in atclBS.findAll("div", attrs={'class': "text"}):
            print(hit.get_text())
```

# Crawling 1: 가장 기본적인 형태의 크롤링/파싱

```python
import pandas as pd

import requests  # 크롤링을 위한 라이브러리
from bs4 import BeautifulSoup  # 파싱을 위한 라이브러리
```

```python
post_df = pd.DataFrame(columns=[
    "검색 키워드",
    "블로그 제목",
    "블로그 이름",
    "블로그 주소",
])
```

 - 일단은 검색할 키워드를 미리 선정합니다.
 - 여러분들이 아마 데이터 분석을 하실 예정이면, 이런 키워드 또한 크롤링, 혹은 어떠한 방식으로 생성하여
 - 크롤링을 수행해야 합니다
 - 이렇게 할 수도 있겠지만, 아래처럼 한번 키워드 리스트를 만들어보겠습니다.

```python
keywords = [
    "신사동 원룸",
    "신사역 원룸",
    "가로수길 원룸",
    "가로수길 오피스텔",
    "신사역 부동산",
]

# 지역 키워드
location_keywords = [
    "신사동",
    "신사역",
    "압구정동",
    "압구정역",
    "가로수길",
]

# 접미사 키워드
postfix_keywords = [
    "부동산",
    "원룸",
    "오피스텔",
    "투룸",
    "방",
]
```

```python
keywords = [
    "{location_keyword} {postfix_keyword}".format(
        location_keyword=location_keyword,
        postfix_keyword=postfix_keyword
    )
    for location_keyword in location_keywords
    for postfix_keyword in postfix_keywords
]

keywords
```

 - https://search.naver.com/search.naver?query=신사동+원룸 형태로 검색 되고 있다. 

```python
response = requests.get("https://search.naver.com/search.naver?query=신사동+원룸")
dom = BeautifulSoup(response.content, "html.parser")
blog_post_elements = dom.select('li.sh_blog_top > dl > dt > a')
#blog_post_elements
blog_post_elements[0].attrs.get('title')
```

```python
blog_post_elements = dom.select('li.sh_blog_top')
blog_post_elements[0].select('a.sh_blog_title')
```

```python
blog_post_elements[0].select('a.sh_blog_title')[0].attrs.get('title')
```

```python
for keyword in keywords:
    # 우리가 작성한 키워드는 " " ( 띄어쓰기 ) 로 되어 있지만,
    # 네이버에서 검색을 할 떄는 " " => "+" 로 변경되서 검색되어야 합니다.
    
    query = keyword.replace(' ', '+')
    response = requests.get("https://search.naver.com/search.naver?query={query}".format(query=query))
    assert response.status_code is 200
    dom = BeautifulSoup(response.content, "html.parser")
    blog_post_elements = dom.select('li.sh_blog_top')
    for blog_post_element in blog_post_elements:
        title = blog_post_element.select('a.sh_blog_title')[0].attrs.get('title')
        blog_name = blog_post_element.select('a.txt84')[0].text
        url = blog_post_element.select('a.sh_blog_title')[0].attrs.get('href')
        
        post_df.loc[len(post_df)] = [keyword, title, blog_name, url]
```
