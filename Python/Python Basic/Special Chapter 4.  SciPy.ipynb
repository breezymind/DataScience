# SciPy

```python
import scipy as sp
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
rv = sp.stats.norm(loc=10, scale=10)
```

```python
rv.rvs(size=(3,10), random_state=1)
```

```python
plt.rcParams['figure.figsize'] = (8, 6)
sns.distplot(rv.rvs(size=10000, random_state=1))
```

```python
import numpy as np
```

```python
xx = np.linspace(-40,60,1000)
pdf = rv.pdf(xx)
```

```python
plt.plot(xx,pdf)
```

```python
cdf = rv.cdf(xx)
```

```python
plt.plot(xx,cdf)
```

# scipy.special 수학 함수

```python
x = np.linspace(-3, 3, 1000)
y1 = sp.special.erf(x)
a = plt.subplot(211)
plt.plot(x,y1)
plt.title("erf")
a.xaxis.set_ticklabels([])
y2 = sp.special.expit(x)
plt.subplot(212)
plt.plot(x, y2)
plt.title("logistic")
```

```python
plt.rcParams['figure.figsize'] = (10, 10)
np.random.seed(0)
tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips)
tips.describe()
```

```python
iris = sns.load_dataset("iris")
sns.jointplot("sepal_width", "petal_length", data=iris, kind="kde", space=0, zorder=0, n_levels=6)
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"], size=2)
```
