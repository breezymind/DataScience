## 변수 
 - 파이썬에서 변수는 객체이다. 파이썬에서 보았던 자료형을 포함한 그 모든 것을 객체라고 한다. 
 - 객체는 메모리에서 생성된다. 
 - x = 4 를 생성하면 4라는 객체가 메모리에서 생성되고 그 객체를 x 가 가리키게 된다. 여기서 x 를 레퍼런스라고 한다. 
 - 여기서 x = 8로 한다면 4 라는 객체는 그대로 존재하고 8 이라는 객체가 다시 만들어지며 x 는 8 객체를 가르키는 레퍼런스가 된다. 
 - 하지만 여기서 4 객체가 그대로 있다면 메모리가 문제가 될 것이다. ( 찌꺼기가 쌓이는 것처럼 ) 하지만 파이썬은 JAVA와 마찬가지로 가비지 컬랙터라는 것이 존재한다. 객체 중 아무런 참조가 없는 객체를 수거해가는 쓰레기차와 비슷한 느낌의 객체이다.

### 변수 할당 및 비교, 삭제 
 - 각의 객체를 할당한 변수(레퍼런스)는 is로 T/F로 출력된다. 
 - del(변수) 를 통해 변수를 삭제할 수 있다. 하지만 이것은 객체를 삭제한다기보단 -> 표를 끊는다는 표현이 어울릴 것 같다.  

```python
a = 3
b = 3
print(a is b)  # 3이라는 객체를 a와 b가 가르키고 있어서 동일하다고 판단한다. 여기서 레퍼런스 카운트는 2이다. 

print(a)
del(a)
print(b)
del(b)
print(a + b)  
```

### 리스트 복사 
 - 같은 객체 복사 
 - 새로운 객체로 만들어서 복사 
 - copy 모듈

```python
#리스트 복사 
a = [1,2,3] 
b = a  # => 같은 객체를 바라 본다.
a[1] = 4    
print(a)
print(b) 

## => 4로 바꾼 뒤 a,b 둘 다 바뀌는 이유는 같은 객체를 바라보고 있기 때문이다. 
```

```python
b = a[:]
print(b)
a[1] = 3
print(a)
print(b)
```

```python
from copy import copy 
b = copy(a)
print(b)

print(a is b)
```

## if 
 - space bar 또는 tab으로 다른 언어의 {} 로 대체 
 - if 조건 : 으로 if 조건문 생성
 - 들여쓰기 길이를 항상 같게 유지해야 한다. 만약 다르다면 인식이 되지 않는다. 

```python
money = 1
if money:  # 조건 후  ' : ' 를 작성 해 주어야 한다. 
    print("택시 타라")
else:
    print("걸어라")
```

```python
money = 100
if money >= 100:
    print("비행기타라")
else:
    print("배타라")
```

```python
watch  = 1

if money > 100 or watch:
    print("오잉")
else:
    print("오잉2")
```

## While 문 
 - 반복해서 문장 실행. 모든언어와 비슷한 메커니즘을 가지고 있다. 
 - while 조건문: 
 - while 참조건문: ( 무한 루프 ) 

```python
treehit = 0
while treehit < 10:
    treehit = treehit + 1
    print("나무를 %d 번 찍었습니다." % treehit)
    if treehit == 10:
        print("쓰러진다.")
```

```python
#while 참조건문: 
    # 무한 루프 
prompt = """
. . . 1. Add
. . . 2. Del
. . . 3. List
. . . 4. Quit
. . .
. . . Enter Number : """
    
number = 0
while number != 4:
    print(prompt)
    number = int(input())  # raw_input => String 형태로 키보드 입력 값을 받아온다. => int로 묵시적 형변환  => python3부터 input()으로통일.
```

### Break 
 - while 문 또는 For 문 과 같이 반복문을 탈출 할 때 사용. 

```python
coffee = 5
money = 300
while money:
    print("커피")
    coffee = coffee - 1
    print("남은 커피 : %d" % coffee) 
    if not coffee:
        print("오링남")
        break;
```

### Continue 
 - 반복문의 처음으로 돌아갈때. 아래를 수행하지 않고 그대로 돌아갈때 

```python
a = 0 
while a<10:
    a = a + 1
    if a % 2 == 0: continue
    print(a)
```

### Pass와 Continue의 차이점. 
 - Pass는 해당 부분을 넘기고 아래까지 진행
 - Continue는 처음으로 돌아간다. 

```python
a = 0 
while a<10:
    a = a + 1
    if a % 2 == 0: pass
    print(a)
```

## For 문
 - 시작과 끝점을 지정해놓고 반복문을 돌리는 문장이다. 
 - R과 비슷하게 문자열을 매개변수로 두고 반복할 수 있다. ( 리스트, 튜플, 문자열 )
 - for 변수 in 리스트:
 - while과 마찬가지로 break, continue 모두 사용 가능하다.

```python
test_list = ['one','two','three']
for i in test_list:
    print(i)
```

```python
marks = [90,25,67,45,80]
number = 0 
for mark in marks:
    number += 1
    if mark >= 60:
        print("%d번째 학생은 합격입니다." % number)
    else:
        print("%d번째 학생은 불합격입니다." % number)
```

```python
number = 0 
for mark in marks:
    number += 1
    if mark >= 60:
        print("%d번째 학생은 합격입니다." % number)
        break
    else:
        print("%d번째 학생은 불합격입니다." % number)
```

### Range 
 - R의 sep 와 동일하다고 보면 된다. 
 - range()   : 0부터 () 갯수만큼 + 1 씩
 - range(시작,끝)   : 시작 ~ 끝 - 1까지 
 - range(시작,끝,갭) : 시작 ~ 끝  - 1까지 갭만큼.

```python
print(range(10))  #  range 객체를 만들어준다. 
print(range(1,10))
print(range(1,10,2))
```

```python
sum = 0 
for i in range(5):
    sum = sum + i
    print("%d 누적 %d " % (i,sum))
print(sum)
```

```python
a = [(1,2),(3,4),(5,6)]
for (first,last) in a:
    print("%d + %d = %d" % (first,last,first+last))
```

```python
marks = [90,25,67,45,80]
for number in range(len(marks)):
    if marks[number] >= 60:
        print("%d번째 학생은 합격입니다." % number)
    else:
        print("%d번째 학생은 불합격입니다." % number)
```

```python
for i in range(2,10):
    for j in range(1,10):
        print("%d * %d = %d" % (i,j,i*j))
```
