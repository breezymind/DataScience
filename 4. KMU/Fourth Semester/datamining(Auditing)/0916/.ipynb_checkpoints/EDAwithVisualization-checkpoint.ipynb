# 데이터 시각화를 통한 탐색적 데이터 분석
탐색적 데이터 분석이란?
 1. 데이터에 대한 질문을 찾는다.
 2. 데이터에 대한 시각화, 변환, 모델링으로 답을 찾는다.
 3. 답을 통해 질문을 심화하고 새로운 질문을 찾는다.

## ggplot2
R에서 가장 널리 쓰이는 시각화 패키지 

```r
library(ggplot2)
```

### 기본 데이터 ( Diamonds )

```r
data(diamonds)
```

```r
head(diamonds)
```

## ggplot의 기초 
 - 공통요소들은 ggplot() 안에 넣고 각 그래프 (geom_이름)에는 개별 속성 적용

## 1. 히스토그램 (Histogram)

```r
options(repr.plot.width=7,repr.plot.height=4) # Jupyter notebook 에서 ggplot 크기를 조절 
```

```r
ggplot(data=diamonds) +
    geom_bar(mapping = aes(x=cut)) # 데이터를 diamonds로 하고 x축에는 cut필드를 기준으로 하여 barplot을 생성
```

```r
library(dplyr)
```

```r
diamonds %>% 
    count(cut)
```

연속 변수의 경우 일정 간격으로 끊어서 히스토그램을 그린다. 아래는 0.5 간격으로 끊어 그림 

```r
ggplot(data = diamonds) +
  geom_histogram(mapping = aes(x = carat), binwidth = 0.5)
```

```r
diamonds %>%
    count(cut_width(carat, .5))
```

```r
smaller <- diamonds %>% filter(carat < 3)
head(smaller)
```

이전과 달리 mapping 내용이 ggplot안에 들어가는 이유는 공통적인 요소라고 생각되어 ggplot 안으로 넣었다.

```r
ggplot(data = smaller, mapping = aes(x = carat)) +
     geom_histogram(binwidth = .1)
```

## 2. Freqploy

여러개의 Histogram 이 겹치면 잘 보이지 않으므로 freqploy를 쓰는 것이 용이 

```r
ggplot(data = smaller, mapping = aes(x = carat, colour = cut)) +
    geom_freqpoly(binwidth = .1)
```

```r
ggplot(data = smaller, mapping = aes(x = carat, colour = cut)) +
    geom_histogram(binwidth = .1, position = )
```

## 다른 변수들에서 분포의 특성을 찾아보기 

```r
ggplot(data = diamonds, mapping = aes(x = carat)) +
     geom_histogram(binwidth = .01)
```

## 3. Outlier(이상점)

```r
ggplot(data = diamonds) +
    geom_histogram(mapping = aes(x = y), binwidth = .5)
```

### 세로축을 제한하여 그리기 

```r
ggplot(diamonds) + 
  geom_histogram(mapping = aes(x = y), binwidth = .5) +
  coord_cartesian(ylim = c(0, 50))
```

```r
unusual <- diamonds %>%
        filter(y < 3 | y > 20) %>%
        select(price, x,y,z) %>%
        arrange(y)
```

```r
head(unusual,10)
```

## 4. Missing Value(결측치)

```r
diamonds2 <- diamonds %>%
    mutate(y = ifelse(y<3 | y>20,NA,y))
```

```r
head(diamonds2)
```

```r
ggplot(data = diamonds2, mapping = aes(x = x, y = y)) + 
  geom_point()
```

```r
ggplot(data = diamonds2, mapping = aes(x = x, y = y)) + 
  geom_point(na.rm = TRUE) # 결측치 제거 
```

## 5. Count vs Density

```r
ggplot(data = diamonds, mapping = aes(x = price)) + 
  geom_freqpoly(mapping = aes(colour = cut), binwidth = 500)
```

```r
ggplot(data = diamonds, mapping = aes(x = price, y = ..density..)) + 
  geom_freqpoly(mapping = aes(colour = cut), binwidth = 500)
```

## 6. Box plot

```r
ggplot(data = diamonds, mapping = aes(x = cut, y=price)) +
    geom_boxplot()
```

```r
ggplot(data = diamonds, mapping = aes(x = cut, y = price)) +
  geom_boxplot() + 
  coord_flip()
```

## 7. 두 개의 이산 변수 

```r
ggplot(data = diamonds) + 
    geom_count(mapping = aes(x = cut, y = color))
```

```r
diamonds %>% 
  count(color, cut) %>%  
  ggplot(mapping = aes(x = color, y = cut)) +
    geom_tile(mapping = aes(fill = n))
```

## 7. 두개의 연속 변수 
 - 점이 겹쳐보기 떄문에 몇개의 점이 있는지 정확히 알 수 없다.

```r
ggplot(data = diamonds) +
  geom_point(mapping = aes(x = carat, y = price))
```

 - 점의 투명도를 통해 많이 있는 부분에는 색이 찐하게 나올 수 있도록 투명도 조절

```r
ggplot(data = diamonds) + 
  geom_point(mapping = aes(x = carat, y = price), alpha = 1 / 100)
```

```r
ggplot(data = smaller) +
  geom_bin2d(mapping = aes(x = carat, y = price))
```
