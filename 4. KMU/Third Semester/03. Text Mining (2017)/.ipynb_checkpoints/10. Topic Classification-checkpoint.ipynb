#### 1.7.1 크롤링

```python
import itertools

import lxml.html
import requests
import tqdm
import re
```

```python
url_base = 'http://stackoverflow.com/questions/tagged/{tag}?page={page}&sort=newest&pagesize=50'
```

```python
tags = ['rstudio', 'dplyr', 'ggplot2']
```

```python
list(itertools.product(tags,range(1,3))) # 조합을 만들어준다.
```

```python
tag_page = list(itertools.product(tags, range(1, 21)))
```

```python
titles = []
title_tag = []
for tag, page in tqdm.tqdm_notebook(tag_page):
    url = url_base.format(tag=tag, page=page)
    res = requests.get(url)
    root = lxml.html.fromstring(res.text)
    for link in root.cssselect('h3 a.question-hyperlink'):
        title = link.text
        title = re.sub(tag, ' ', title, flags=re.IGNORECASE) # 태그에 해당되는 부분은 공백으로 변경 => 예측이 쉽기 때문에.
        titles.append(title)
        title_tag.append(tag)
```

```python
len(titles) # 50개 글씩 60번 반복 => 3000개
```

```python
titles[:3]
```

```python
len(title_tag)
```

```python
title_tag[:3]
```

#### 1.7.2 데이터 준비.

##### TDM

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english', max_features=1000)
```

```python
tdm = cv.fit_transform(titles)
```

##### train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(tdm, title_tag, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(y_train)
```

```python
len(y_test)
```

#### 1.7.3 Logistic Regression
##### Training 
 - 기본 LogisticRegression 으로 하게 되면 안의 계수들을 찾아서 한다. 힘들기 대문에 LogisticRegressionCV(Cross Validation) 을 넣게 되면 알아서 데이터를 쪼개서 검증용이랑 학습용으로 나눠 해본다. 

```python
from sklearn.linear_model import LogisticRegressionCV 
```

```python
logreg = LogisticRegressionCV(Cs=100)
```

```python
logreg.fit(X_train,y_train)
```

##### 결과 

```python
logreg.C_ # 알파의 역수를 알려준다. RStudio, dplyr, ggplot 따로따로 예측.
```

```python
logreg.classes_
```

```python
logreg.score(X_train,y_train) ## Accuracy 
```

```python
logreg.score(X_test,y_test) ## Accuracy 
```

#### 평가 

```python
y_logreg = logreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
logreg.classes_
```

```python
metrics.confusion_matrix(y_test, y_logreg)  # row: true, col: predicted
```

```python
metrics.accuracy_score(y_test, y_logreg)
```

```python
metrics.f1_score(y_test, y_logreg, average=None)
```

 - Deep Learning의 경우 하나하나가 의미가 없을때는 Deep Learning이 좋다( MRI로 암 예측)

### 2.1.1 장점
 - 이해하기 쉽다.
 - 전처리가 단순
 - 빠르다. 
 - 다양한 종류의 변수를 다룰수 있다.
 - 모형의 시각화
 - 통계적 가정이 적다.
 
### 2.1.2 단점
 - 과적합(overfitting)
 - 결과의 불안정
 - 최적화가 어려움
 - 학습시키기 어려운 문제들이 있음(예: XOR)
 - Imbalanced data에 취약
  - 극단적으로 몰린 데이터. -> 다수의 경우를 예측하면 학습이 없더라도 맞출 수 있다. 
  - 질문의YES/NO차이가 많은걸 찾는데 다 차이가 없어진다.


### 2.2 앙상블(Ensemble)
 - 여러 모델을 만들어서 각 모델들을 다수결을 시킨다던지 하는 방식
 - 하나의 모형은 under-/over-fitting 될 수 있음 
 - 앙상블: 여러 개의 모형을 만들어 다수결/평균을 사용
 - 배깅(bagging 또는 bootstrap aggregation):
  - 데이터에서 일부 변수의 샘플을 무작위로 뽑는다
  - 샘플에 모형을 학습시킨다 (각각을 Decision Tree)
  - 1-2를 반복하여 여러 개의 모형을 만든다
  - 위의 모형들의 예측의 다수결/평균으로 예측한다

 - Random Forest: DT + bagging
![img31](img/31.PNG)

### 2.3 부스팅(boosting)
 - 모든 데이터에 동일한 가중치 
 - 데이터로 모형 1 학습
 - 모형 1이 틀린 데이터의 가중치를 높임 -> 1번은 Rstudio를 잘맞추면 2번은 RStudio는 빼고 학습, 2번은 ggplot2를 못맞춘다 그러면 ggplot2에 집중하여. 
 - 데이터로 모형 2를 학습. 
 - 위를 반복. 
 - 1번이 만들어질때까지 기다려야 된다. 그래서 시간이 오래 걸린다. 

### 2.4 경사 부스팅(Gradient Boosting)
 1. 데이터로 모형1을 학습
 2. 모형1의 예측과 실제의 오차
 3. 위의 오차를 모형2를 학습
 4. 1-3의 과정을 반복



 - 실제값 = 모형1의 예측 + 모형1의 오차
 - 모형1의 오차 = 모형2의 예측 + 모형2의 오차
 - 모형2의 오차 = 모형3의 예측 + 모형3의 오차
 - 실제값 = 모형1의 예측 + 모형2의 예측 + … + 아주 작은 오차



<a href="https://xgboost.readthedocs.io/en/latest/"> xgboost (Scalable and Flexible Gradient Boosting) </a>

### 실습. Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier(n_estimators=10, random_state=0)
```

```python
gb.fit(X_train, y_train)
```

```python
y_gb = gb.predict(X_test.todense())  # .todense()
```

```python
metrics.confusion_matrix(y_test, y_gb)  # row: true, col: predicted
```

```python
metrics.accuracy_score(y_test, y_gb)
```

### Gradient Boosting with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
gb = GradientBoostingClassifier()
```

```python
cv = GridSearchCV(gb, {'n_estimators': [10, 20], 'max_depth': [1, 3]})
```

```python
cv.fit(X_train.toarray(), y_train)
```

```python
cv.best_params_
```

```python
y_gbcv = cv.predict(X_test.toarray())
```

```python
metrics.confusion_matrix(y_test, y_gbcv)  # row: true, col: predicted
```

```python
metrics.accuracy_score(y_test, y_gbcv)
```
