```python
url
```

```python
texts = []
for page in range(1, 1000):
    res = requests.get(url.format(page))
    element = lxml.html.fromstring(res.content)
    for text in element.xpath('.//div[@class="ct"]'):
        texts.append([
                text.find('.//a[@class="tit"]').text,
                text.find('.//span[@class="time"]').text
            ])
```

```python
texts
```

```python
import csv
```

```python
with open('통일인식변화.csv', 'w', encoding='utf8') as f:
    writer = csv.writer(f)
    writer.writerows(texts)
```

```python
with open('통일인식변화.csv', encoding='utf8', newline='\r\n') as f:
    reader = csv.reader(f)
    texts = list(reader)
```
