# 한 글자 짜리 및 불필요한 단어 빼기

```python
exclude=['에서','경제','전망','이후','기자','올해']
```

```python
def get_word(doc):
    nouns = tagger.nouns(doc)
    return [noun for noun in nouns if len(noun) > 1 and noun not in exclude]
```

```python
cv = CountVectorizer(tokenizer=get_word, max_features=50)
```

```python
tdm = cv.fit_transform(news)
```

```python
words = cv.get_feature_names()
```

# 단어간 상관계수

```python
word_corr = numpy.corrcoef(tdm.todense(), rowvar=0)
```

```python
word_corr
```

# 상관관계 높은 것 100개만 추리기

```python
edges = []
for i in range(len(words)):
    for j in range(i + 1, len(words)):
        edges.append((words[i], words[j], word_corr[i, j]))
```

```python
edges = sorted(edges, key=operator.itemgetter(2), reverse=True)
```

2; 상관계수

```python
edges = edges[:100]
```

```python
edges
```

상관성이 있는 단어를 보면 미 연방과 연준, 연준과 금리 인상, 경제 상황과 통화정책, 기준금리는 인하할 것인지에 대한 관심, 
수출이 회복될 것인지, 국내 투자가 살아날 것인지, 한국은행 총재는 금리를 인하할 것인가, 경기에 따른 통화정책은 어떻게, 
소비와 수출이 살아나야 하는데, 수출 부진이면 성장은 정체하고 경기는 하향 등의 주제가 거론된 것으로 판담됨

```python
edge_list = [(word1, word2) for word1, word2, weight in edges]
weight_list = [weight for word1, word2, weight in edges]
```

# 상관관계 시각화

```python
import networkx
```

```python
G = networkx.Graph()

edge_set = set()
for word1, word2, weight in edges:
    G.add_edge(word1, word2, weight=weight)
```

```python
position = networkx.spring_layout(G, iterations=30)
pyplot.figure(figsize=(12, 9))
networkx.draw_networkx_nodes(G, position, node_size=0)
networkx.draw_networkx_edges(G, position, edgelist=edge_list, width=weight_list, edge_color='pink')
networkx.draw_networkx_labels(G, position, font_size=15, font_family='Malgun Gothic')
pyplot.axis('off')
pyplot.show()
```

금리 수준이 가장 중심에 위치하여 인하에 대한 관심과 경기에 끼치는 영향, 세계 경제와의 관계, 통화정책 등으로 연결됨을 알 수 있고,
금리나 통화정책, 경기 영향, 그리고 세계 경제라는 키워드 등에 관련된 경제 요인들이 연결됨을 알수 있음
