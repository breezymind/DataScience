# 스크립트 언어

 - 스크립트 언어란? 스크립트 ( Script ) 의 사전적 의미 - 연극의 대사 등이 적혀 있는 것에서 유래
 - a programming language that supports scripts, programs written for a special run-time environment that can interpret (rather than    compile) and automate the execution of tasks that could alternatively be executed one-by-one by a human operator, line by line 
 - 소스코드 = 스크립트
 - 연기자가 스크립트를 보고 연기 = 컴퓨터가 스크립트를 읽어 수행 
 - a special run-time environment = 수행시간 환경 
 - 해석과 수행을 동시에 함
 - 즉, 라인 by 라인으로 해석하여 수행.

# 문자열 치환과 포메팅 
 - 문자열 치환과 포메팅
 - 문자열 정렬과 공백
 - 소수점 표현
 - 아래와 같은 방법으로 %s 문자열, %c 문자 한개, %d 숫자, %f 실수 ( 부동소수 ) 를 대입할 수 있다.
 - 해당 변수의 데이터 타입을 모를 경우는 %s 표현하면 모든 부분에 있어서 표현이 가능하다.
 - 하지만 이와 같이 코딩을 하게 된다면 좋지 않은 프로그램으로 판단 될 것이다. 변환작업이 일어나야 하기 때문이다. 

```python
# 문자열 치환 
a = "pithon"
print(a[1])
#a[1] = 'y'  #에러가 난다. 
a[:1] + 'y' + a[2:]

# 문자열 포메팅 C언어와 비슷한 형태.
print('i eat %d apples.' % 3)
cnt = 4
print('i eat %d apples.' % cnt)

print('i eat %d%% apples.' % 3)         # %를 표현해 주기위해선 %% 2개가 필요하다.

# 정렬과 공백
print("%10s" % "hi") #왼쪽에서 10개인 문자열 공간에서 오른쪽 정렬. 
print("%-10s" % "hi") #왼쪽에서 10개인 문자열 공간에서 왼쪽 정렬. 

# 소수점 표현
print("%0.4f" % 3.234555)   #소수점 4번째까지만 표현 '.' 앞의 숫자는 위와 마찬가지로 정렬을 어디로 할지 정하는 부분이다.
```

# Dictionary
 - Key Value 형태로 존재한다.
 - List와 Tuple 과 달리 index으로 구하는 것이 아니라. key로 value를 구한다. 
 - {} 로 표혀한다. List [] Tuple () Dictionary {}

```python
dict = {}
dict['one'] = "This is one"
dict[2]     = "This is two"

tinydict = {'name': 'john','code':6734, 'dept': 'sales'}


print (dict['one'])       # Prints value for 'one' key
print (dict[2])           # Prints value for 2 key
print (dict)
print (tinydict)          # Prints complete dictionary
print (tinydict.keys())   # Prints all the keys
print (tinydict.values()) # Prints all the values
```

### Dictionary 쌍 추가, 삭제 
 - 새로운 key, value를 설정하여 바로 추가 할 수 있다. 
 - 추가되는 순서는 원칙이 없다. 중요한 것은 무엇이 어떤 값을 가지고 추가 됐는지다. 
 - Key 는 고유한 값으로 중복되는 값을 설정해 놓으면 하나를 제외한 나머지의 것들은 무시된다는 점이다. 
   무엇이 무시될지는 예측이 불가능하다. 즉, Key를 중복되게 사용하지 마라. 
 - Key 에는 리스트는 사용할 수 없다. 하지만 튜플은 가능하다. 이유는 리스트와 딕셔너리는 내용이 변형이 가능하지만 튜플은 고정형으로 데이터가 변형이 안되므로 상수 역할이 가능하다. 

```python
a ={1:'a'}
a[2] = 'b'
print(a)
a['name'] = 'Lee'
print(a)
a[3] = [1,2,3]
print(a) # 정렬되면서 입력이 되는 것 같다. 숫자부터 문자열로 정렬되면서.

del a[1]
print(a)
del a['name']
print(a)

a = {[1,2] : 'b'}
```

### Dictionary 함수 
 - key List 만들기
 - Value List 만들기
 - Key, Value 쌍 만들기
 - Key, Value 쌍 모두 지우기. 
 - Key 로 Value 얻기 ( Get ) 
 - 해당 Key 유무 확인. ( has_key )

```python
a = { 'name' : 'Lee', 'phone' : '01012341234', 'birth' : 881108 }
print(a.keys())  # Key List
print(a.values()) # Value List
print(a.items()) # Key , Value 쌍 얻기.
print(a.get('name'))  # Key로 Value get
print(a.get('phone'))  # Key로 Value get
#print(a.has_key('name'))  # Key 유무   python 2.7에서 사용가능
#print(a.has_key('tel'))

print('name' in a)
print('phones' in a)

a.clear() # Clear
print(a.items())
```

### set : 집합형 자료형
 - python 2.3부터 지원 
 - 집합형태의 자료들을 처리하기 위해 만들어짐.
 - set 키워드를 사용.

```python
s1 = set([1,2,3])
s1
```

```python
s2 = set("hello")
s2
```

#### 위와 같은 특징을 가지는 이유 
 - 순서가 없다.
 - 중복을 허용하지 않는다. 
 - 순서가 없으므로 인덱싱을 할 수가 없다. 
 - 인덱싱으로 하려면 형변환이 필요하다. 

```python
s1 = set([1,2,3])
l1 = list(s1)
l1[0]
```

```python
t1 = tuple(s1)
t1[0]
```

#### 활용 
 - 교집합 
 - 합집합
 - 차집합

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
```

- 교집합

```python
print(s1 & s2)
print(s1.intersection(s2))
```

- 합집합 

```python
print(s1 | s2)
print(s1.union(s2))
```

- 차집합

```python
print(s1 - s2)
print(s1.difference(s2))
```

#### 관련 함수 
 - 값 추가 (add : 한개 update : 여러개 )
 - 값 제거 (remove : 삭제 )

```python
s1.add(4)
s1
```

```python
s1.update([7,8,9])
s1
```

```python
s1.remove(4)
s1
```
