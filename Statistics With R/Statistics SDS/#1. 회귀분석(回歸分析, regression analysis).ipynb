# 회귀분석(回歸分析, 영어: regression analysis)


```r
employees <- read.csv("data//hfda_ch10_employees_kr.csv", stringsAsFactors=F)
head(employees)
```

```r
options(repr.plot.width=4, repr.plot.height=3)
hist(employees$RECEIVED, breaks=50)
```

```r
sd(employees$RECEIVED)
```

```r
summary(employees$RECEIVED)
```

```r
options(repr.plot.width=10, repr.plot.height=5)
par(mfrow = c(2,3))
hist(employees$RECEIVED[employees$YEAR==2007], breaks=50, ann=F)
title("year 2007 ", xlab="RECEIVED")
hist(employees$RECEIVED[employees$YEAR==2008], breaks=50, ann=F)
title("year 2008 ", xlab="RECEIVED")
hist(employees$RECEIVED[employees$SEX=="F"], breaks=50, ann=F)
title("FEMALE ", xlab="RECEIVED")
hist(employees$RECEIVED[employees$SEX=="M"], breaks=50, ann=F)
title("MALE ", xlab="RECEIVED")
hist(employees$RECEIVED[employees$NEGOTIATED==F], breaks=50, ann=F)
title("Negotiated False ", xlab="RECEIVED")
hist(employees$RECEIVED[employees$NEGOTIATED==T], breaks=50, ann=F)
title("Negotiated True ", xlab="RECEIVED")
```

```r
plot(employees$REQUESTED[employees$NEGOTIATED==T],employees$RECEIVED[employees$NEGOTIATED==T])
```

## 상관계수 
 - 두개의 변수의 관계의 강도.
 - 공분산 / 두개의 분산 곱 
 - 공분산은 부호 ( 양인지 음인지 )

```r
cor(employees$REQUESTED[employees$NEGOTIATED==T],employees$RECEIVED[employees$NEGOTIATED==T])
```

```r
aLm <- lm(RECEIVED[NEGOTIATED==T]~ REQUESTED[NEGOTIATED==T], data = employees)
plot(employees$REQUESTED[employees$NEGOTIATED==T],employees$RECEIVED[employees$NEGOTIATED==T])
abline(aLm)
```

### 통계적 분석 오류처리
 - 회귀식 범위내의 데이터는 예측이 가능하지만 그 의외의 값에서는 예측하기가 힘들다. ( 불가능 )
 - 외삽 : 데이터 붐위 밖의 값을 예측 하는 것. 
 - 보간 : 데이터 범위 내에서 값을 예측 하는 것, 이것이 회귀를 사용하는 목적. 

 - 확률 오차 (change error) : 모델이 예측한 것과 결과의 편차 
 - 잔차라고도 표현.

### R.M.S 
 - 오차를 사용해 잔차 분포를 정량화 한다. 
 - 오차 공식
  - 정량적으로 나타내거나 사용될 수 있다.
  - 항상 예측에 대한 오차를 설명 할 수 있어야 한다.

```r
summary(aLm)$sigma
```
