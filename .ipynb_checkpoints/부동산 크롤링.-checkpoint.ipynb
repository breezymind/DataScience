```python
import pandas as pd
import html5lib
```

```python
url = "http://realestate.daum.net/iframe/maemul/maemulList.daum?areaCode=3406737&mcateCode=A1A3A4&saleTypeCode=*&tabName=maemulList&isExpanded=false&page=1"
```

```python
dfs = pd.read_html(url,encoding="utf-8")
```

```python
len(dfs)
```

```python
df = dfs[1] 
print(len(df), 'rows')
```

```python
df.head(10)
```

```python
df = df.dropna()
print(len(df), 'rows')
```

```python
df.head(10)
```

```python
df1 = df.ix[::2,:]

print(len(df1), 'rows')
```

```python
df1.head(10)
```

```python
df2 = df.ix[1::2, 0]
df2.name = '비고'
df2.head(10)
```

```python
# df1과 df2를 합치기(concat)위해 위해 index를 동일하게 초기화 한다.
df1.reset_index(drop=True , inplace=True)
df2.reset_index(drop=True , inplace=True)

# df1 과  df2 를 컬럼으로 합치기
result = pd.concat([df1, df2], axis=1)
result.head(10)
```

```python
result['거래종류'] = result['최초게재일'].str.extract('(.*) ')
result.head(10)
```

```python
result['최초게재일'] = result['최초게재일'].str.extract('(\d+.\d+.\d+)')
result.head(10)
```
