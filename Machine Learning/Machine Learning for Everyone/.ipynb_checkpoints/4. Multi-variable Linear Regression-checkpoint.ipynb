# Multi-variable Linear Regression

```python
import tensorflow as tf
import numpy as np
```

### Data set

```python
x1_data = [1,0,3,0,5]
x2_data = [0,2,0,4,0]
y_data = [1,2,3,4,5]
```

### find values for W and b that comput y_data = W * x_data + b

```python
W1 = tf.Variable(tf.random_uniform([1],-1.0,1.0)) # -1.0 ~ 1.0의 난수
W2 = tf.Variable(tf.random_uniform([1],-1.0,1.0)) # -1.0 ~ 1.0의 난수
b = tf.Variable(tf.random_uniform([1],-1.0,1.0))
```

### Hypothesis

```python
hypothesis = W1 * x1_data + W2 * x2_data + b
```

### Simplified cost function
 - $\sum_{i=1}^m(W_x^{(i)}-y^{(i)})^2$

```python
cost = tf.reduce_mean(tf.square(hypothesis-y_data))
```

### Minimize 

```python
a = tf.Variable(0.1)  # learning rate, alpha
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)  # goal is minimize cost
```

### Initialize the variables & Launch

```python
init = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init)

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W1),sess.run(W2),sess.run(b))
```

```python
W = tf.Variable(tf.random_uniform([1, len(x_data)], -1, 1)) # len(x_data) -> [1,3] 으로 생성하여 마지막에 b와 x_data의 1열과 행렬곱
b = tf.Variable(tf.random_uniform([1], -1, 1))
```

 - matmul : Matrix Multiplication

```python
hypothesis = tf.matmul(W, x_data)
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_data))

a = tf.Variable(0.1)  # learning rate, alpha
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)  # goal is minimize cost

init = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init)

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W))
```
