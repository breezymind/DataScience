# Pivot_Table and Group_by, Cut

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data\salesfunnel.xlsx')
df.head()
```

```python
df.info()
```

```python
df['Status'].unique()
```

```python
df['Status'] = df['Status'].astype('category')
df['Status'].cat.set_categories(["won",'pending','presented','declined'], inplace=True)
```

```python
df.head()
```

```python
pd.pivot_table(df,index=['Name']) # 숫자료 표현된 것들은 전부 mean 이 적용 된다. 
```

```python
pd.pivot_table(df,index=['Manager','Rep']) #  각각을 groupby하여 숫자형을 mean
```

 - Value를 선택할 수도 있다. 

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price']) #mean적용.
```

 - 평균으로 표현하던 것을 다른 것으로 변경 하고 싶다면 aggfunc 속성에 함수를 넣으면 된다. 

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'], aggfunc=np.sum) #np.sum적용.
```

 - 혹은 2가지로 표현도 가능하다. 

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'], aggfunc=[np.mean,len])
```

 - columns=['Product']로 각 제품들을 표현할 수도 있다. 
 - fill_value를 통해 NaN값을 치환할 수 있다. 또는 다 된 값에 fillna(값)으로 치환 가능하다. 

```python
#pd.pivot_table(df,index=['Manager','Rep'],values=['Price'], columns=['Product'], aggfunc=[np.sum]).fillna(0)
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'], columns=['Product'], aggfunc=[np.sum], fill_value=0)
```

```python
pd.pivot_table(df,index=['Manager','Rep'], values=['Price','Quantity'], columns=['Product'], aggfunc=[np.sum], fill_value=0)
```

```python
pd.pivot_table(df,index=['Manager','Rep','Product'], values=['Price','Quantity'],aggfunc=[np.sum],fill_value=0)
```

 - 소숫점 제거 : margins = False

```python
pd.pivot_table(df,index=['Manager','Rep','Product'], values=['Price','Quantity'],aggfunc=[np.sum,np.mean],fill_value=0, margins=False)
```

 - 소수점이 길어진다. 

```python
pd.pivot_table(df,index=['Manager','Rep','Product'], values=['Price','Quantity'],aggfunc=[np.sum,np.mean],fill_value=0, margins=True)
```

 - True로 두면 나눗셈 연산을 수행한 경우 소숫점이 길어지고 전체 항목을 잡아줍니다.
 - 평균이면 평균, 합계면 합계로 말이죠. 소숫점이 길어진 것도 평균에서 평균을 잡으니 한 columns의 소숫점 길이가 길어졌기 때문입니다.


```python
pd.pivot_table(df,index=['Manager','Status'], values=['Price'], aggfunc=[np.sum], fill_value=0, margins=True) # True를 할경우 총합이 나오네
```

 - aggfunc에 적용한 컬럼명을 지정 할 수도 있다. 

```python
pd.pivot_table(df, index=['Manager','Status'], columns=['Product'], 
               values=['Quantity','Price'], 
               aggfunc={"Quantity":len, "Price":np.sum},
              fill_value=0)
```

 - 두 개의 values에 하나는 크기(len)만 적용했지만, Price에는 np.sum과 np.mean 이렇게 두 함수를 적용한 경우

```python
table = pd.pivot_table(df, index=['Manager','Status'], columns=['Product'], 
               values=['Quantity','Price'], 
               aggfunc={"Quantity":len, "Price":[np.sum,np.mean]},
              fill_value=0)
table
```

### query를 이용한 조건 문. 

```python
table.query('Manager==["Debra Henley"]')
```

```python
table.query('Status==["pending","won"]')
```

## GroupBy 

```python
raw_data = { 'regiment' : ['Nighthawks','Nighthawks','Nighthawks','Nighthawks'
                          ,'Dragoons','Dragoons','Dragoons','Dragoons','Scouts',
                           'Scouts','Scouts','Scouts'],
            'company':['1st','1st','2nd','2nd',
                       '1st','1st','2nd','2nd',
                       '1st','1st','2nd','2nd'],
            'name':['Miller','Jacobson','Ali','Milner','Cooze','Jacon','Ryaner','Sone','Sloan','Piger','Riani','Ali'],
            'preTestScore' : [4,24,31,2,3,4,24,31,2,3,2,3],
            'postTestScore' : [25,94,57,62,70,25,94,57,62,70,62,70]}
raw_data
```

```python
print(len(raw_data['regiment']))
print(len(raw_data['company']))
print(len(raw_data['name']))
print(len(raw_data['preTestScore']))
print(len(raw_data['postTestScore']))
```

```python
df = pd.DataFrame(raw_data)
```

```python
df.head()
```

```python
groupby_regiment = df['preTestScore'].groupby(df['company'])
```

```python
groupby_regiment
```

```python
list(groupby_regiment)
```

```python
df['preTestScore'].groupby(df['company']).describe()
```

```python
groupby_regiment.mean()
```

```python
df['preTestScore'].groupby([df['regiment'],df['company']]).mean()
```
