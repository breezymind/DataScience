# 베이즈 이론 

## 1.1 조건부 확률 
 - 베이지안 통계에 기본 개념이 바로 베이즈 이론이다.
 - 확률은 명제 혹은 예측에 대한 확신도를 0~1사이로 표현한다. 
  - 1 : 참이거나 실제로 일어났음.
  - 0 : 거짓이거나 일어나지 않음.
 - 조건부 확률 
  - $P(A|B)$ : B라는 조건이 주어졌을 때의 A가 참일 확률.
  - 예) A는 내가 내년에 심장발작을 일으키는 것, B는 나머지 조건 이다. 

## 1.2 결합 확률
 - A와 B가 모두 참인 확률 : $P(A \ and \ B)$
 - $P(A \ and \ B)$ = $P(A)$ x $P(B)$, 단 A,B가 독립일 경우에만 이다. 
  - 예) 동전이 앞면이 나올 확률이라고 할때 $P(A \ and \ B)$ = 0.5 x 0.5 = 0.25
  - 즉, 첫번째 사건의 결과를 안다고 해도 두 번째 사건의 확률이 바뀌지 않아야 한다는 것. 
  - $P(B|A)$ = $P(B)$
 - 독립이 아닐 경우에는 
  - A는 오늘 비 올확률. B는 내일 비가 올 확률이라고 한다면, 오늘 비가 왔다는 것을 알고 있다면 내일도 비가 올 확률이 높다. 
  - $P(B|A)$ > $P(B)$
  - $P(A \ and \ B)$ = $P(A)P(B|A)$

## 1.3 쿠기 문제
 - 첫번째 그릇, 바닐라 30, 초코 10
 - 두번째 그릇, 바닐라, 초코 20 20 
 - 바닐라 쿠키가 그릇 첫번째에서 나왔을 확률은? 
  - $P(vanilla \ | \ Bowl 1)$

## 1.4 베이즈 이론
 - 교환 가능하다는 조건
 - $P(A \ and \ B)$ = $P(B \ and \ A)$
 - $P(A \ and \ B)$ = $P(A) \ P(B|A)$
 - $P(B \ and \ A)$ = $P(B) \ P(A|B)$
 - $P(A)P(B|A)$ = $P(B) \ P(A|B)$ 라는 조건이 만족한다. 
 
 $$P(A|B) = {P(A)P(B|A) \over P(B)}$$
 
 - 위의 쿠키 문제 
  - 1번 그릇에서 꺼냇을 경우 $B_1$, 바닐라 쿠키 $V$
  - $P(B_1|V) = {P(B_1)P(V|B_1) \over P(V)}$
  - $P(B_1|V) = {P(1/2)P(3/4) \over P(5/8)}$

## 1.5 통시적 해석
데이터 D의 관점에서 봤을 때 가설 H의 확률을 수정해주는 식이다. 이러한 방법을 통시적 해석이라고 한다. 
 - 통시적 은 무언가가 시간에 따라 일어나는 것으로, 가설에 대한 확률이 시간에 따라 새로운 데이터를 접하게 되며서 달라진다는 뜻이다. 

 $$P(H|D) = {P(H)P(D|H) \over P(D)}$$
  - $P(H)$ : **사전확률**(데이터 보기 전 확률)
  - $P(H|D)$ : 데이터를 확인한 이후의 가설로, **사후 확률**이라고 한다.
  - $P(D|H)$ : 데이터가 가설에 포함될 확률, **우도(가능도)**라고 한다. 
  - $P(D)$ : 어떤 가설에든 포함되는 데이터의 비율, **한정 상수**라고한다.
  
 보통 가정 집합을 단순화하여 정의 한다. 이러한 성격의 가설 집합을 스윗(Suite)이라고 한다.
  
  
  - **상호 배제 **
   - 집합 중 하나의 가설만 참일 경우 
  - ** 전체 호괄 **
   - 다른 가능성이 전혀 없는 경우. 단 하나의 가설이라도 참일 경우 
   
 $P(D)$ = $P(B_1)P(D|B_1)$ + $P(B_2)P(D|B_2)$ => $(1/2)(3/4)$  + $(1/2)(1/2)$ = $5/8$

## 1.6 M&M 문제 

Mars 사는 1995년 M&M에 파란색을 추가. <br>
그전에는 기본 봉지안에 색조합은 **갈색 30%, 노랑 20%, 빨강 20%, 녹색 10%, 주황 10%, 황갈색 10%** 였다.<br>
파란색을 추가한 이후로 **파랑 24%, 녹 20%, 주 16%, 노 14%, 빨 13%, 갈 13%**가 되었다.<br>

두 봉지를 샀을때 각각 생산 년도가 94, 96년 이었다. 하나씩 꺼냈을 때 한 알은 노란색이고, 한 알은 녹색이었다.<br>
이때 노랑 초콜렛이 1994년에 생산한 봉지에서 나왔을 확률은?

### 1.6.1 가설의 수치화 
 - 노란색 초콜렛 : 봉지 1
 - 녹색 초콜렛 : 봉지 2에서 꺼냈다고 하자. 
 
 A : 봉지 1 - 1994, 봉지 2 - 1996 <br>
 B : 봉지 1 - 1996, 봉지 2 - 1994
 
|   | 사전확률 | 우도     |              | 사후 확률 |
|---|----------|----------|--------------|-----------|
|   | $P(H)$   | $P(D|H)$ | $P(H)P(D|H)$ | $P(H|D)$  |
| A | 1/2      | (20)(20) | 200          | 20/27     |
| B | 1/2      | (10)(14) | 70           | 7/27      |


- 사전 확률 : A, B 모두 1/2로 추정할 수 있다. 
- 우도 
 - A가 참이라면 노랑 20%(1994), 녹색 20%(1996), 독립 결과 
 - B가 참이라면 노랑 14%(1996), 녹색 10%(1994), 독립 결과 
- 세번째 열 
 - 앞의 두 열의 곱 
- 마지막은 세번째 열을 확률로 구한 것.

## 1.7 몬티홀 문제 
 - 거래를 하자라는 게임 쇼의 사회자
 - 세 개의 문을 보여주고 뒤에 상품이 있다고 말할 것이다.비싼자동차, 나머지 두개는 땅콩 버터나 가짜 손톱 같은 비교적 저가의 상품이다. 상품은 임의로 놓여있다.
 - 게임 목적은 뒤에 자동차가 있는 문을 맞추는 것이다. 맞춘다면 자동차를 가질 수 있다.
 - A,B,C 문이 있고 A문을 선택 했다고 하자. 
 - 선택한 문을 열기 전에, 몬티는 B,C 중에 자동차가 없는 문을 열어 긴장감을 높인다. 자동차는 세 문 중 한곳에만 있으니 B,C중 하나는 분명 자동차가 없을 것이다. 
 - 그 다음 몬티는 이제 그대로 문 A를 선택할지 아니면 열리지 않은 다른 문으로 바꿀지 다시 선택하라고 한다. 
 
 
대부분 차이가 없을 것으로 판단, 문은 두개이므로 차가 문 A뒤에 있을 확률이 50%라고 추측. <br>
사실 문 A를 고수했을 떄 맞출 확률은 1/3일 뿐이지만, 문을 바꾸면 맞출 확률은 2/3이다. <br>
데이터를 명확히 하자 D는 두 부분으로 구성된다. 몬티는 B를 선택했고 거기에 자동차가 없다는 것이다. <br>
A,B,C는 각 문 뒤에 차가 있다는 가설이다.

|   | 사전확률 | 우도     |              | 사후 확률 |
|---|----------|----------|--------------|-----------|
|   | $P(H)$   | $P(D|H)$ | $P(H)P(D|H)$ | $P(H|D)$  |
| A | 1/3      | 1/2      | 1/6          | 1/3       |
| B | 1/3      | 0        | 0            | 0         |
| C | 1/3      | 1        | 1/3          | 2/3       |


 - 각 문에 차가 있을 확률은 조건이 없으므로 1/3 로 동일하다.
 - 만약 차가 실제로 문 A뒤에 있다면 몬티는 문 B/C를 안전하게 열 수 있다. 따라서 몬티가 문 B를 선택할 확률은 1/2이다. 그리고 A뒤에 있으므로 차가 문 B 뒤에 없을 확률은 1이다. 
 - 마지막 차가 문 C 뒤에 있다면 몬티는 1의 확률로 문 B를 열것이고, 1의 확률로 차가 없을 것이다. 
 - $P(A|D)$=1/3 
 - $P(C|D)$=2/3
