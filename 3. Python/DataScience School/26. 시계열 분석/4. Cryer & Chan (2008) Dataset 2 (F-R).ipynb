## `flow`

* Flow data (in cubic feet per second) for the Iowa River measured at Wapello, Iowa
* 09/1958–08/2006
* Source: http://waterdata.usgs.gov/ia/nwis/sw

```python
df = pd.read_csv("data/tsa/flow.dat")
df.index = pd.date_range("09/1958", "08/2006", freq="MS")
df.head()
```

```python
df.plot()
```

## `gold`

* Daily price of gold (in U.S. dollars per troy ounce)

* Source: www.lbma.org.uk/2005dailygold.htm

```python
df = pd.read_csv("data/tsa/gold.dat")
df.index = pd.date_range("01/04/2005", periods=len(df), freq="B")
df.head()
```

## `google`

* Daily returns of Google stock
* 08/20/04 to 09/13/06. 
* Source: http://finance.yahoo.com/q/hp?s=GOOG

```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

df = pd.read_csv("data/tsa/google.dat")
df.index = pd.date_range("08/20/04", periods=len(df), freq=CustomBusinessDay(calendar=USFederalHolidayCalendar())) 
df.tail()
```

```python
df.plot()
```
