## airpass
 - Monthly total international airline passengers
 - 01/1960-12/1971
 - Source: Box, G. E. P., Jenkins, G. M., and Reinsel, G. C.,Time Series Analysis: Forecasting and Control, second edition, Prentice-Hall, Engelwood Cliffs, NJ, 1994.

```python
df = pd.read_csv("Datasets/airpass.dat", header=None)
df.index = pd.date_range("01/1960", "12/1971", freq="MS")
df.columns = ["airpass"]
```

```python
df.head()
```

```python
df.plot()
```

## bluebird 
 - Weekly unit sales of Bluebird standard potato chips (New Zealand) and their price for 104 weeks
 - Source: www.stat.auckland.ac.nz/~balemi/Assn3.xls

```python
df = pd.read_csv('Datasets/bluebird.dat', sep=' ')
```

```python
df.tail()
```

```python
plt.subplot(211)
df["log.sales"].plot()
plt.subplot(212)
df.price.plot()
```
