# 결정론적 추세 추정 
## 추세(Trend)
 - 확률 과정의 기대값이 시간  $t$ 에 대한 함수로 표현될 수 있으면 이를 추세($trend$)라고 한다.
 - $f(t)$ 가 상수가 아닌 $t$에 따라 변하는 경우, 추세성을 띈다라고 한다. 

```python
import numpy as np
import scipy.stats as sp 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(0)
t = np.arange(20)
x = 2 * t + sp.norm.rvs(size=20)
plt.plot(t, x, 'o-')
plt.show()
```

```python
np.random.seed(0)
t = np.arange(20)
x = -5*np.sin(0.25*np.pi*t) + sp.norm.rvs(size=20)
plt.plot(t, x, 'o-')
plt.show()
```
