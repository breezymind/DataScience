# Data Distribution : 데이터 분포 
 - 데이터가 어떠한 모양으로 모여있는지 전반적인 형태를 살피기 위해서 사용하는 것. 
 - 일차원 데이터의 분포를 보기 위해서는

## 1. 기술 통계 (descriptive statistics)
 - 데이터의 숫자 (count)
 - 평균 (mean, average)
 - 분산 (variance)
 - 표준 편차 (standard deviation)
 - 최댓값 (maximum)
 - 최솟값 (minimum)
 - 중앙값 (median)
 - 사분위수 (quartile)

## 2. Histogram
 - matplotlib 패키지의 hist
 - seaborn 패키지의 distplot
 
 - matplotlib 패키지의 hist 함수는 다음과 같은 3개의 값을 반환한다. 
  - n : 각 구간에 포함된 값의 갯수 혹은 빈도 리스트
  - bins : 구간의 경계값 리스트
  - patches : 각 구간을 그리는 matplotlib patch 객체 리스트

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
%matplotlib inline
```

```python
n, bins, patches = plt.hist(x, bins=10)
```

```python
n
```

```python
bins
```

```python
patches
```

seaborn 패키지의 displot명령은 히스토그램에 대한 axis 객체만을 반환하는 대신 러그(rug), 커널 밀도(kernel density) 등을 표시하거나 특정한 확률 모형으로 fitting하는 추가 기능이 있다.

```python
plt.figure(figsize=(6,4))
sns.distplot(x, rug=True)
```

```python
sns.distplot(x, rug=False)
```

## 3. 다차원 데이터 분포를 묘사하는 경우
 - 만약 2차원 데이터이고 각 데이터가 모두 연속적인 실수값이라면 스캐터 플롯(scatter plot)을 사용하면 된다. 
 - 스캐터 플롯을 그리기 위해서는 seaborn 패키지의 jointplot 명령을 사용한다. 
  - jointplot 명령은 스캐터 플롯뿐 아니라 각 변수의 히스토그램도 동시에 그린다.

```python
iris = sns.load_dataset("iris")
iris.tail()
```

```python
iris.describe() # 각각의 기술통계 출력.
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
```

 - 인수를 바꾸면 커널 밀도의 형태로도 표시할 수 있다.

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde", space=0, zorder=0, n_levels=6)
```

 -  3차원 이상의 데이터라면 seaborn 패키지의 pairplot 명령을 사용한다. pairplot은 그리도(grid) 형태로 각 집합의 조합에 대해 히스토그램과 스캐터 플롯을 그린다.
 - R의 psych library => pairs panels

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"], size=3)
```

 - 카테고리(category) 값이면 seaborn 패키지의 heatmap 명령을 사용
 - 아래는 년, 월별 항공 탑승객의 수를 나타낸 것이다.

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.tail()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(flights, annot=True, fmt="d", linewidths=1, annot_kws={"size":15})
```

 - 하나는 연속적인 값이고 다른 하나는 이산적인 혹은 카테고리 값인 경우에는 seaborn에서 제공하는 다음과 같은 플롯을 사용할 수 있다.
  - boxplot
  - violinplot
  - stripplot
  - swarmplot
  - pointplot
  - factorplot

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

 - 승객의 나이(age) 분포를 성(sex), 선실(class), 기항지(embark_town)에 따라 분류

```python
sns.factorplot(x="age", y="embark_town", hue="sex", row="class", data=titanic[titanic.embark_town.notnull()],
               orient="h", size=3, aspect=3.5, palette="Set3", kind="violin", split=True, cut=0, bw=.2)
```

```python
sns.factorplot(x="age", y="embark_town", hue="sex", row="class", data=titanic[titanic.embark_town.notnull()],
               orient="h", size=3, aspect=3.5, palette="Set3", kind="box")
```
