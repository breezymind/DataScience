# 내장 함수 
 - 파이썬 내장함수 
 - 사용도에 따라 사용 하면 될 듯
 - abs
 - divmod
 - enumerate
 - lambda
 - map
 - filter 

```python
# 절대값 
print(abs(-10))

# divmod(a,b)  a / b (몫,나머지) => tuple
print(divmod(10,3))

print(chr(97))  # 아스키 코드 97번 

#enumerate 
#index와 value를 순서대로 가지고 온다.
for i,name in enumerate(['boby','foo','bee']):  
    print(i,name)  

#lambda

sum = lambda x, y : x + y   # 함수를 쉽게 표현 할 수 있다. x,y를 받아서 return x + y
print(sum(3,4))

# map
# r = map(func, seq)  #seq값을 함수에 적용. 결과를 R 로 

def fahrenheit(T):
    return ((float(9)/5)*T + 32)

def celsius(T):
    return (float(5)/9)*(T-32)

temperatures = (36.5, 37, 37.5, 38, 39)
F = list(map(fahrenheit, temperatures))
C = list(map(celsius, F))
print(F)
print(C)


C = [36.5, 37, 37.5, 38, 39]
F = list(map(lambda x: (float(9)/5)*x + 32, C))
print(F)
C = list(map(lambda x: (float(5)/9)*(x-32), F))
print(C)


#filter 
def positive(l): 
    result = [] 
    for i in l: 
        if i > 0: 
            result.append(i)   # result(list) 에 값 추가.
    return result

print(positive([1,-3,2,0,-5,6]))


def positive2(x):
    return x > 0

print(list(filter(positive2, [1, -3, 2, 0, -5, 6])))  # positive2의 참 값만 걸어서 묶어서 return 



# id 객체 고유 번호 (레퍼런스 번호)
a = 3
print(id(3))

# isinstance (object, class)

class Person: pass

a = Person()
print(isinstance(a,Person))

# len
a = [1,2,3,4,5]
print(len(a))

a = "1234556"
print(len(a))

# list("string")
# "string" to list

print(list("string"))


# max, min 
# 숫자, 문자 의 최대 최소값을 Return

# pow(x,y)  x ^ y 

print(pow(10,2))

def test(x,y):
    return x+y
def test2(x,y):
    return x*y


print(sorted([3,1,2]))
print(sorted(["b","a","c","f","d"]))

# str() 문자형으로 변환 

print(str('abc').upper())
print(str('ABC').lower())

# zip 동일한 갯수의 요소값을 갖는 시퀀스 자료형을 묶어주는 역할을 한다. 

print(list(zip([1,2,3],[4,5,6])))
print(tuple(zip([1,2,3],[4,5,6])))
```

# Thread (쓰레드) 
 - 동작하고 있는 프로그램을 Process 
 - 하나의 Process는 하나의 Work를 한다. 
 - 하나의 Process가 여러개의 Work을 할 수 있도록 하는 것이 Thread이다. 

```python
import threading

class Messenger(threading.Thread):
    def run(self):
        for _ in range(5000):
            print(threading.currentThread().getName())
        
send = Messenger(name="sending out messages")
receive = Messenger(name="receving messages")

send.start()
receive.start()
```
